```python
# %%capture
# Instalaci√≥n de librer√≠as
!pip install yfinance
!pip install pandas_datareader
!pip install ipywidgets

# Importaci√≥n de bibliotecas
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
import yfinance as yf
from scipy.stats import norm
from datetime import datetime, date
import ipywidgets as widgets
from IPython.display import display, Markdown
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import fsolve  # Para c√°lculo de œÉ*

# Suprimir warnings de yfinance
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning)

# Configuraci√≥n inicial
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("colorblind")
pd.options.display.float_format = '{:,.4f}'.format

# Configuraci√≥n de estilo para gr√°ficos
plt.rcParams.update({
    'figure.figsize': (12, 6),
    'font.size': 12,
    'axes.titlesize': 16,
    'axes.titleweight': 'bold',
    'axes.labelsize': 14,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'grid.alpha': 0.3
})

# Paleta de colores corporativa USACH
usach_colors = ['#0D3B66', '#1D7A8C', '#A5C882', '#E56B6F', '#9B5DE5']
sns.set_palette(usach_colors)

# =============================================================================
# TABLA DE FUENTES DE DATOS
# =============================================================================
print("="*80)
print("FUENTES DE DATOS Y SUPUESTOS")
print("="*80)

# Fecha actual para todo el an√°lisis
fecha_actual = date(2025, 6, 19)
print(f"Fecha de an√°lisis: {fecha_actual.strftime('%Y-%m-%d')}\n")

fuentes_datos = pd.DataFrame({
    'Dato': ['Precios de acciones (MSFT)',
             '√çndice de mercado (S&P 500)',
             'Tasa libre de riesgo (10Y)',
             'Precio spot petr√≥leo (WTI)',
             'Precio spot oro'],
    'Fuente': ['Yahoo Finance (yfinance)',
               'Yahoo Finance (^GSPC)',
               'U.S. Department of Treasury (4.25% al 18/06/2025)',
               'NYMEX (Cushing, OK WTI) - Real',
               'London Gold Market Fixing Ltd - Real'],
    'Fecha Extracci√≥n': [fecha_actual.strftime('%Y-%m-%d')]*5,
    'Periodo Datos': ['2018-01-01 a 2025-06-18',
                      '2018-01-01 a 2025-06-18',
                      'Valor actual',
                      'Valor real',
                      'Valor real'],
    'Supuesto/Real': ['Real', 'Real', 'Real', 'Real', 'Real']
})

print("\nFuentes de datos utilizadas en la valoraci√≥n:")
print(fuentes_datos.to_string(index=False))

# Funci√≥n para imprimir encabezados de secci√≥n
def print_header(title, level=1):
    decorator = '=' * 80 if level == 1 else '-' * 60
    print(f"\n{decorator}")
    print(title.upper())
    print(f"{decorator}\n")

# =============================================================================
# 1. VALORACI√ìN DE BONOS (MEJORADA)
# =============================================================================
print_header("1. valoraci√≥n de bonos mejorada", level=1)

"""
Teor√≠a:
La valoraci√≥n de bonos se basa en el valor presente de los flujos futuros (cupones y principal).
La duraci√≥n de Macaulay mide la sensibilidad del precio a cambios en las tasas de inter√©s,
mientras que la convexidad captura la curvatura de esta relaci√≥n (Hull, 2018, Cap. 4).
"""

class ValoradorBono:
    def __init__(self, valor_nominal, tasa_cupon, a√±os, rendimiento_requerido, frecuencia=2):
        self.valor_nominal = valor_nominal
        self.tasa_cupon = tasa_cupon
        self.a√±os = a√±os
        self.rendimiento_requerido = rendimiento_requerido
        self.frecuencia = frecuencia
        self.n = a√±os * frecuencia
        self.cupon = (tasa_cupon * valor_nominal) / frecuencia
        self.tasa_periodo = rendimiento_requerido / frecuencia

    def precio(self):
        """Calcula el precio del bono"""
        vp_cupones = self.cupon * (1 - (1 + self.tasa_periodo)**-self.n) / self.tasa_periodo
        vp_principal = self.valor_nominal / (1 + self.tasa_periodo)**self.n
        return vp_cupones + vp_principal

    def duracion_macaulay(self):
        """Calcula la Duraci√≥n de Macaulay"""
        precio = self.precio()
        flujos = [self.cupon] * self.n
        flujos[-1] += self.valor_nominal
        t = np.array([(i+1)/self.frecuencia for i in range(self.n)])
        vp_flujos = np.array([cf / (1 + self.tasa_periodo)**(i+1) for i, cf in enumerate(flujos)])
        return np.sum(t * vp_flujos) / precio

    def convexidad(self):
        """Calcula la convexidad del bono"""
        precio = self.precio()
        flujos = [self.cupon] * self.n
        flujos[-1] += self.valor_nominal
        t = np.array([(i+1)/self.frecuencia for i in range(self.n)])
        convexidad = 0
        for i in range(self.n):
            factor_descuento = 1 / (1 + self.tasa_periodo)**(i+1)
            periodo = t[i]
            convexidad += flujos[i] * periodo * (periodo + 1) * factor_descuento
        return convexidad / (precio * (1 + self.tasa_periodo)**2)

    def ytm_from_price(self, precio_mercado):
        """Calcula el YTM dado un precio de mercado (m√©todo de Newton-Raphson)"""
        # Estimaci√≥n inicial
        ytm = self.rendimiento_requerido
        max_iter = 100
        tolerancia = 1e-6

        for i in range(max_iter):
            self.rendimiento_requerido = ytm
            self.tasa_periodo = ytm / self.frecuencia
            precio_calculado = self.precio()
            derivada = -self.duracion_macaulay() * precio_calculado / (1 + ytm/self.frecuencia)

            # Ajuste de YTM
            error = precio_calculado - precio_mercado
            if abs(error) < tolerancia:
                break

            ytm -= error / derivada

        return ytm

# Ejemplo pr√°ctico
bono_params = {
    'valor_nominal': 1000,
    'tasa_cupon': 0.06,
    'a√±os': 5,
    'rendimiento_requerido': 0.07,
    'frecuencia': 2
}

bono = ValoradorBono(**bono_params)
precio_bono = bono.precio()
duracion = bono.duracion_macaulay()
convexidad = bono.convexidad()

# Calcular YTM inverso
precio_mercado = 950
ytm_calculado = bono.ytm_from_price(precio_mercado)

# Tabla de resultados
tabla_bono = pd.DataFrame({
    'M√©trica': ['Valor Nominal', 'Tasa Cup√≥n', 'Plazo (a√±os)', 'YTM', 'Precio', 'Duraci√≥n (a√±os)', 'Convexidad'],
    'Valor': [bono_params['valor_nominal'],
              f"{bono_params['tasa_cupon']*100:.2f}%",
              bono_params['a√±os'],
              f"{bono_params['rendimiento_requerido']*100:.2f}%",
              f"${precio_bono:.2f}",
              f"{duracion:.2f}",
              f"{convexidad:.2f}"]
})

print("Resultados de valoraci√≥n de bono:")
print(tabla_bono.to_string(index=False))
print(f"\nYTM calculado para precio de mercado ${precio_mercado}: {ytm_calculado*100:.4f}%")

# =============================================================================
# MARKDOWN 1: BONO - PRECIO Y RIESGO
# =============================================================================
display(Markdown("""
## 1 | Bono ‚Äì Precio y riesgo (Duraci√≥n & Convexidad)

### üîç Qu√© mirar en los dos gr√°ficos del bono

1. **Precio vs YTM (izq.)**
   * La curva cae cuando sube la tasa.
   * La pendiente en torno a ¬±0,25 % se parece a
     \(-\text{Duraci√≥n} √ó Precio\).
   * Si te alejas m√°s, la l√≠nea real se ‚Äúencorva‚Äù: esa diferencia la corrige la **convexidad**.

2. **Duraci√≥n vs YTM (der.)**
   * A tasas bajas la duraci√≥n es larga ‚Üí el bono es muy sensible.
   * A tasas altas la duraci√≥n cae ‚Üí el bono es menos sensible.

üìå **Por qu√© importa**
Duraci√≥n = rapidez con la que cambia el precio; Convexidad = cu√°nto se curva esa relaci√≥n. Juntas explican el 1.¬∫ y 2.¬∫ orden del riesgo de tasa.
"""))

# Gr√°ficos de sensibilidad
fig, axs = plt.subplots(1, 2, figsize=(18, 6))
plt.figure(figsize=(18, 6))
plt.suptitle("Sensibilidad del Precio del Bono al Rendimiento", fontsize=16, fontweight='bold')
plt.savefig("grafico_sensibilidad.png")
plt.close()

# Sensibilidad precio-YTM
rendimientos = np.linspace(0.01, 0.15, 30)
valores = [ValoradorBono(bono_params['valor_nominal'],
                         bono_params['tasa_cupon'],
                         bono_params['a√±os'], r).precio() for r in rendimientos]

axs[0].plot(rendimientos, valores, linewidth=2.5, color=usach_colors[0])
axs[0].axvline(x=bono_params['rendimiento_requerido'], color='r', linestyle='--', alpha=0.7, label='YTM original')
axs[0].axvline(x=ytm_calculado, color='m', linestyle='--', alpha=0.7, label='YTM mercado')
axs[0].axhline(y=precio_bono, color='g', linestyle='--', alpha=0.7, label='Precio calculado')
axs[0].set_title("Sensibilidad del Precio del Bono al Rendimiento", fontsize=14, fontweight='bold')
axs[0].set_xlabel("Rendimiento Requerido (YTM)")
axs[0].set_ylabel("Valor del Bono ($)")
axs[0].grid(True, alpha=0.3)
axs[0].yaxis.set_major_formatter('${x:,.0f}')
axs[0].legend()
axs[0].annotate(f'Precio calculado: ${precio_bono:.2f}\nDuraci√≥n: {duracion:.2f} a√±os\nConvexidad: {convexidad:.2f}',
            xy=(0.05, 1100),
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

# Duraci√≥n vs YTM
rendimientos_duracion = np.linspace(0.01, 0.20, 30)
duraciones = [ValoradorBono(bono_params['valor_nominal'],
                         bono_params['tasa_cupon'],
                         bono_params['a√±os'], r).duracion_macaulay() for r in rendimientos_duracion]

axs[1].plot(rendimientos_duracion, duraciones, linewidth=2.5, color=usach_colors[1])
axs[1].axvline(x=bono_params['rendimiento_requerido'], color='r', linestyle='--', alpha=0.7, label='YTM original')
axs[1].axhline(y=duracion, color='g', linestyle='--', alpha=0.7, label='Duraci√≥n actual')
axs[1].set_title("Sensibilidad de la Duraci√≥n al Rendimiento", fontsize=14, fontweight='bold')
axs[1].set_xlabel("Rendimiento Requerido (YTM)")
axs[1].set_ylabel("Duraci√≥n (a√±os)")
axs[1].grid(True, alpha=0.3)
axs[1].legend()
axs[1].annotate(f'Duraci√≥n actual: {duracion:.2f} a√±os',
            xy=(0.15, 3.5),
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.show()
plt.savefig("grafico_sensibilidad.png")

# =============================================================================
# 2. VALORACI√ìN DE ACCIONES - MEJORADA
# =============================================================================
print_header("2. valoraci√≥n de acciones mejorada", level=1)

"""
Teor√≠a:
El modelo de descuento de dividendos (DDM) valora acciones como el valor presente
de sus dividendos futuros. El CAPM (Capital Asset Pricing Model) se utiliza para
determinar la tasa de descuento adecuada (Gitman & Joehnk, 2015, Cap. 8):

E(R) = Rf + Œ≤ √ó (E(Rm) - Rf)

Donde:
E(R) = Retorno esperado (tasa de descuento)
Rf   = Tasa libre de riesgo
Œ≤    = Beta (sensibilidad al mercado)
E(Rm) = Retorno esperado del mercado
(E(Rm) - Rf) = Prima de riesgo de mercado (ERP)
"""

def descargar_datos_seguro(tickers, inicio='2018-01-01', fin=fecha_actual.strftime('%Y-%m-%d')):
    """Descarga datos hist√≥ricos de forma segura usando yfinance"""
    try:
        # Descargar con auto_adjust=False para evitar warnings
        data = yf.download(tickers, start=inicio, end=fin, progress=False,
                          group_by='ticker', auto_adjust=False)
        return data
    except Exception as e:
        print(f"Error al descargar datos: {str(e)}")
        return None

# Definir los activos a descargar
tickers = ["MSFT", "^GSPC"]

print(f"Descargando datos para {tickers}...")
data_completo = descargar_datos_seguro(tickers)

if data_completo is not None and not data_completo.empty:
    # Extraer precios de cierre para cada activo
    precios = pd.DataFrame()

    for t in tickers:
        if t in data_completo:
            # Manejar estructura de datos diferente para un solo activo vs m√∫ltiples
            if len(tickers) == 1:
                precios[t] = data_completo['Close']
            else:
                precios[t] = data_completo[t]['Close']

    if precios.empty:
        print("Error: No se encontraron datos de precios")
    else:
        # Renombrar columnas para claridad
        precios.columns = ['Accion', 'Indice_Mercado']
        precios = precios.dropna()

        if precios.empty:
            print("Error: No hay datos coincidentes entre la acci√≥n y el √≠ndice")
        else:
            # =================================================================
            # 2.1 Datos
            # =================================================================
            print_header("2.1 datos hist√≥ricos", level=2)

            # Mostrar primeros registros
            print("\nPrimeros 5 registros de precios:")
            print(precios.head().to_string(float_format='${:,.2f}'.format))

            # =================================================================
            # 2.2 M√©tricas de riesgo
            # =================================================================
            print_header("2.2 m√©tricas de riesgo", level=2)

            # Calcular retornos diarios
            retornos = precios.pct_change().dropna()

            # Calcular m√©tricas de riesgo
            retorno_anual = retornos['Accion'].mean() * 252
            volatilidad_anual = retornos['Accion'].std() * np.sqrt(252)
            sharpe = retorno_anual / volatilidad_anual if volatilidad_anual != 0 else 0

            # Calcular retornos semanales
            retornos_semanales = precios.resample('W').last().pct_change().dropna()
            retorno_semanal = retornos_semanales['Accion'].mean()
            volatilidad_semanal = retornos_semanales['Accion'].std()

            # Calcular retornos mensuales
            retornos_mensuales = precios.resample('M').last().pct_change().dropna()
            retorno_mensual = retornos_mensuales['Accion'].mean()
            volatilidad_mensual = retornos_mensuales['Accion'].std()

            # Coeficiente de variaci√≥n
            coef_variacion_diario = volatilidad_anual / retorno_anual if retorno_anual != 0 else np.nan
            coef_variacion_semanal = volatilidad_semanal / retorno_semanal if retorno_semanal != 0 else np.nan
            coef_variacion_mensual = volatilidad_mensual / retorno_mensual if retorno_mensual != 0 else np.nan

            # Calcular beta usando covarianza
            covarianza = retornos.cov().iloc[0, 1]
            varianza_mercado = retornos['Indice_Mercado'].var()
            beta = covarianza / varianza_mercado if varianza_mercado != 0 else np.nan

            # Tabla de m√©tricas
            tabla_metricas = pd.DataFrame({
                'M√©trica': ['Retorno Anual', 'Volatilidad Anual', 'Ratio de Sharpe',
                           'Retorno Semanal', 'Volatilidad Semanal',
                           'Retorno Mensual', 'Volatilidad Mensual',
                           'Coef. Variaci√≥n Diario', 'Coef. Variaci√≥n Semanal', 'Coef. Variaci√≥n Mensual'],
                'Valor': [f"{retorno_anual*100:.2f}%",
                          f"{volatilidad_anual*100:.2f}%",
                          f"{sharpe:.4f}",
                          f"{retorno_semanal*100:.2f}%",
                          f"{volatilidad_semanal*100:.2f}%",
                          f"{retorno_mensual*100:.2f}%",
                          f"{volatilidad_mensual*100:.2f}%",
                          f"{coef_variacion_diario:.4f}" if not np.isnan(coef_variacion_diario) else "N/A",
                          f"{coef_variacion_semanal:.4f}" if not np.isnan(coef_variacion_semanal) else "N/A",
                          f"{coef_variacion_mensual:.4f}" if not np.isnan(coef_variacion_mensual) else "N/A"]
            })

            print("\nM√©tricas de riesgo para MSFT:")
            print(tabla_metricas.to_string(index=False))

            # =================================================================
            # 2.3 Valoraci√≥n con CAPM y DDM
            # =================================================================
            print_header("2.3 valoraci√≥n con capm y ddm", level=2)

            # F√≥rmula CAPM
            display(Markdown("**Modelo CAPM:**"))
            display(Markdown(r"$E(R) = R_f + \beta \times (E(R_m) - R_f)$"))

            # Widgets para ajustar par√°metros CAPM
            print("\nAjuste de par√°metros CAPM:")

            def actualizar_capm(rf, erp):
                tasa_descuento = rf + beta * erp
                print(f"\nTasa de descuento calculada: {tasa_descuento*100:.2f}%")
                return tasa_descuento

            # Valores iniciales (Damodaran ERP 2025 = 5.5%)
            # Actualizado: Tasa libre de riesgo al 18/06/2025 = 4.25%
            tasa_libre_riesgo = 0.0425
            prima_mercado = 0.055
            tasa_descuento = actualizar_capm(tasa_libre_riesgo, prima_mercado)

            try:
                # Solo mostrar widgets si estamos en un entorno interactivo
                widgets.interact(actualizar_capm,
                               rf=widgets.FloatSlider(value=0.0425, min=0.01, max=0.10, step=0.001, description='Rf:'),
                               erp=widgets.FloatSlider(value=0.055, min=0.03, max=0.10, step=0.001, description='ERP:'))
                print("\nFuente: Damodaran (2025) - ERP US = 5.5%")
                print("Fuente Rf: U.S. Department of Treasury - 10-Year Bond Yield (18/06/2025)")
            except:
                print("\nPara widgets interactivos, ejecutar en Jupyter Notebook")

            # Valoraci√≥n con DDM
            # Datos reales de dividendos de Microsoft (2023)
            ult_dividendo = 3.00  # Dividendo anual de Microsoft actualizado
            tasa_crecimiento = 0.08  # Estimaci√≥n de crecimiento de dividendos

            # Validar si la tasa de descuento es mayor que la tasa de crecimiento
            if tasa_descuento <= tasa_crecimiento:
                print("Advertencia: La tasa de descuento debe ser mayor que la tasa de crecimiento para el modelo DDM")
                valor_ddm = np.nan
            else:
                valor_ddm = ult_dividendo * (1 + tasa_crecimiento) / (tasa_descuento - tasa_crecimiento)

            # Resultados en tabla
            tabla_valoracion = pd.DataFrame({
                'M√©trica': ['Beta',
                           'Tasa Libre de Riesgo', 'Prima de Mercado', 'Tasa Descuento (CAPM)',
                           '√öltimo Dividendo', 'Tasa Crecimiento', 'Valor DDM'],
                'Valor': [f"{beta:.4f}" if not np.isnan(beta) else "N/A",
                          f"{tasa_libre_riesgo*100:.2f}%",
                          f"{prima_mercado*100:.2f}%",
                          f"{tasa_descuento*100:.2f}%",
                          f"${ult_dividendo:.2f}",
                          f"{tasa_crecimiento*100:.2f}%",
                          f"${valor_ddm:.2f}" if not np.isnan(valor_ddm) else "N/A"]
            })

            print("\nResultados de valoraci√≥n:")
            print(tabla_valoracion.to_string(index=False))
            print("\nNota: El modelo DDM no considera recompras de acciones, lo que puede explicar diferencias con el precio de mercado")

            # =================================================================
            # MARKDOWN 2: ACCI√ìN MSFT - PRECIO, RIESGO Y MERCADO
            # =================================================================
            display(Markdown("""
            ## 2 | Acci√≥n MSFT ‚Äì Precio, riesgo y mercado

            ### üîç Qu√© mirar en los cuatro gr√°ficos de MSFT

            * **Precio logar√≠tmico** ‚Äì la l√≠nea recta en log-scale indica crecimiento compuesto constante.
            * **Histograma** ‚Äì la anchura muestra la volatilidad; la cola derecha cuenta subidas fuertes.
            * **Volatilidad m√≥vil** ‚Äì picos en 2020 (COVID) y 2022 (inflaci√≥n); te dice cu√°ndo el activo fue m√°s riesgoso.
            * **Dispersi√≥n vs S&P-500** ‚Äì la pendiente es Œ≤ ‚âà 1,18 ‚Üí MSFT se mueve un 18 % m√°s que el √≠ndice.

            üìå **Idea clave**
            Sharpe 0,94: el mejor equilibrio retorno-riesgo dentro de las clases de activo analizadas.
            """))

            # =================================================================
            # 2.4 Sensibilidad y Gr√°ficos
            # =================================================================
            print_header("2.4 an√°lisis de sensibilidad", level=2)

            # Widgets para an√°lisis de sensibilidad DDM
            def actualizar_ddm(g, k):
                if k <= g:
                    print("Error: k debe ser mayor que g")
                    return
                valor = ult_dividendo * (1 + g) / (k - g)
                print(f"\nValor estimado con g={g*100:.1f}% y k={k*100:.1f}%: ${valor:.2f}")

            try:
                widgets.interact(actualizar_ddm,
                                g=widgets.FloatSlider(value=tasa_crecimiento, min=0.01, max=0.15, step=0.01, description='g:'),
                                k=widgets.FloatSlider(value=tasa_descuento, min=0.01, max=0.25, step=0.01, description='k:'))
            except:
                print("\nPara widgets interactivos, ejecutar en Jupyter Notebook")

            # Gr√°ficos
            fig, axs = plt.subplots(2, 2, figsize=(16, 12))

            # Precios hist√≥ricos (escala logar√≠tmica)
            axs[0, 0].plot(precios['Accion'], linewidth=1.5, color=usach_colors[0])
            axs[0, 0].set_yscale('log')
            axs[0, 0].set_title(f"Evoluci√≥n de Precios: MSFT (Escala Log)", fontweight='bold')
            axs[0, 0].set_ylabel("Precio ($ - Log Scale)")
            axs[0, 0].grid(True, alpha=0.3)
            axs[0, 0].yaxis.set_major_formatter('${x:,.0f}')

            plt.savefig("grafico_precios.png")

            # Distribuci√≥n de retornos
            sns.histplot(retornos['Accion'], kde=True, ax=axs[0, 1], bins=50, color=usach_colors[1])
            mean_ret = retornos['Accion'].mean()
            std_ret = retornos['Accion'].std()
            axs[0, 1].axvline(mean_ret, color='r', linestyle='dashed', linewidth=1)
            axs[0, 1].axvline(mean_ret + std_ret, color='g', linestyle='dashed', linewidth=1)
            axs[0, 1].axvline(mean_ret - std_ret, color='g', linestyle='dashed', linewidth=1)
            axs[0, 1].set_title("Distribuci√≥n de Retornos Diarios", fontweight='bold')
            axs[0, 1].set_xlabel("Retorno Diario")
            axs[0, 1].grid(True, alpha=0.3)
            axs[0, 1].text(0.05, 0.95, f"Media: {mean_ret:.4f}\nDesv. Est√°ndar: {std_ret:.4f}",
                           transform=axs[0, 1].transAxes, verticalalignment='top',
                           bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

            plt.savefig("grafico_distribucion.png")

            # Rolling volatility
            volatilidad_rolling = retornos['Accion'].rolling(21).std() * np.sqrt(252)
            axs[1, 0].plot(volatilidad_rolling, linewidth=1.5, color=usach_colors[2])
            axs[1, 0].set_title("Volatilidad Anualizada (Rolling 21 d√≠as)", fontweight='bold')
            axs[1, 0].set_ylabel("Volatilidad")
            axs[1, 0].grid(True, alpha=0.3)
            axs[1, 0].yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

            plt.savefig("grafico_volatilidad.png")

            # Relaci√≥n con mercado
            sns.regplot(x=retornos['Indice_Mercado'], y=retornos['Accion'], ax=axs[1, 1],
                        scatter_kws={'alpha':0.5, 'color':usach_colors[3]},
                        line_kws={'color':usach_colors[4]})
            axs[1, 1].set_title(f"Relaci√≥n MSFT vs S&P 500", fontweight='bold')
            axs[1, 1].set_xlabel("Retorno S&P 500")
            axs[1, 1].set_ylabel("Retorno MSFT")
            axs[1, 1].grid(True, alpha=0.3)
            if not np.isnan(beta):
                axs[1, 1].text(0.05, 0.05, f"Beta = {beta:.4f}", transform=axs[1, 1].transAxes,
                               bbox=dict(facecolor='white', alpha=0.8))

            plt.savefig("grafico_relacion.png")

            plt.tight_layout()
            plt.show()

else:
    print("Error: No se pudieron descargar datos v√°lidos para el an√°lisis de acciones")

# =============================================================================
# DESCARGAR PRECIOS REALES DE MATERIAS PRIMAS
# =============================================================================
print_header("actualizaci√≥n de precios spot reales", level=1)

# Descargar precios reales para WTI y Oro
try:
    # Precio spot WTI (CL=F es el s√≠mbolo de Yahoo Finance para el futuro inmediato de WTI)
    datos_wti = yf.download("CL=F", start=fecha_actual, end=fecha_actual)
    if not datos_wti.empty:
        spot_wti = datos_wti['Close'].iloc[0]
        print(f"Precio spot WTI actualizado: ${spot_wti:.2f}")
    else:
        spot_wti = 80.0
        print("Advertencia: No se pudo descargar el precio spot WTI. Se mantiene valor supuesto.")

    # Precio spot Oro (GC=F es el s√≠mbolo de Yahoo Finance para el futuro inmediato de oro)
    datos_oro = yf.download("GC=F", start=fecha_actual, end=fecha_actual)
    if not datos_oro.empty:
        spot_oro = datos_oro['Close'].iloc[0]
        print(f"Precio spot oro actualizado: ${spot_oro:.2f}")
    else:
        spot_oro = 1800.0
        print("Advertencia: No se pudo descargar el precio spot oro. Se mantiene valor supuesto.")
except Exception as e:
    print(f"Error al descargar precios spot: {e}")
    spot_wti = 80.0
    spot_oro = 1800.0

# =============================================================================
# 3. VALORACI√ìN DE FUTUROS (MEJORADA)
# =============================================================================
print_header("3. valoraci√≥n de futuros mejorada", level=1)

"""
Teor√≠a:
El precio te√≥rico de un futuro se calcula considerando el precio spot,
los costos de almacenamiento, y el rendimiento de conveniencia (Hull, 2018, Cap. 5):

F = S √ó e^{(r + c - y)T}

Donde:
F = Precio futuro
S = Precio spot
r = Tasa libre de riesgo
c = Costo de almacenamiento
y = Rendimiento de conveniencia
T = Tiempo hasta vencimiento
"""

class ValoradorFuturos:
    def __init__(self, precio_spot, tasa_libre_riesgo, tiempo, costo_almacenamiento=0, rendimiento_conveniencia=0):
        self.precio_spot = precio_spot
        self.tasa_libre_riesgo = tasa_libre_riesgo
        self.tiempo = tiempo
        self.costo_almacenamiento = costo_almacenamiento
        self.rendimiento_conveniencia = rendimiento_conveniencia

    def precio_teorico(self):
        """Calcula el precio te√≥rico de un contrato de futuro"""
        return self.precio_spot * np.exp(
            (self.tasa_libre_riesgo + self.costo_almacenamiento - self.rendimiento_conveniencia) * self.tiempo
        )

    def tabla_sensibilidad(self, tasas=None, costos=None):
        """Genera tabla de sensibilidad"""
        if tasas is None:
            tasas = np.linspace(0.01, 0.10, 5)
        if costos is None:
            costos = np.linspace(0.00, 0.10, 5)

        resultados = []
        for t in tasas:
            for c in costos:
                self.tasa_libre_riesgo = t
                self.costo_almacenamiento = c
                precio = self.precio_teorico()
                resultados.append({
                    'Tasa': t,
                    'Costo Almacen': c,
                    'Precio Futuro': precio
                })

        return pd.DataFrame(resultados)

# Ejemplo: Futuro de petr√≥leo (WTI)
futuro_params = {
    'precio_spot': spot_wti,  # Precio real actualizado
    'tasa_libre_riesgo': 0.0425,  # Actualizado a tasa actual
    'tiempo': 6/12,
    'costo_almacenamiento': 0.02,
    'rendimiento_conveniencia': 0.01
}

valorador_futuro = ValoradorFuturos(**futuro_params)
futuro_teorico = valorador_futuro.precio_teorico()

# Tabla de sensibilidad
sensibilidad_tasas = np.linspace(0.02, 0.08, 4)
sensibilidad_costos = np.linspace(0.00, 0.05, 4)
tabla_sensibilidad = valorador_futuro.tabla_sensibilidad(sensibilidad_tasas, sensibilidad_costos)

# Tabla de resultados
tabla_futuros = pd.DataFrame({
    'Par√°metro': ['Precio Spot', 'Tasa Libre Riesgo', 'Tiempo (a√±os)',
                  'Costo Almacenamiento', 'Rend. Conveniencia', 'Precio Futuro Te√≥rico'],
    'Valor': [f"${futuro_params['precio_spot']:.2f}",
              f"{futuro_params['tasa_libre_riesgo']*100:.2f}%",
              f"{futuro_params['tiempo']:.2f}",
              f"{futuro_params['costo_almacenamiento']*100:.2f}%",
              f"{futuro_params['rendimiento_conveniencia']*100:.2f}%",
              f"${futuro_teorico:.2f}"]
})

print("\nResultados de valoraci√≥n de futuros:")
print(tabla_futuros.to_string(index=False))

print("\nTabla de sensibilidad:")
print(tabla_sensibilidad.pivot(index='Tasa', columns='Costo Almacen', values='Precio Futuro').to_string(float_format='${:,.2f}'.format))

# =============================================================================
# MARKDOWN 3: FUTURO WTI - MODELO COST-OF-CARRY
# =============================================================================
display(Markdown("""
## 3 | Futuro WTI ‚Äì Modelo cost-of-carry

### üîç Qu√© mirar

* **Precio futuro vs plazo** ‚Äì l√≠nea ascendente ‚áí ‚Äúcontango‚Äù: sostener el barril cuesta dinero.
* **Convergencia al spot** ‚Äì al acercarse la fecha el futuro debe acercarse al spot, eliminando arbitraje.
* **Mapa de calor** ‚Äì cada n√∫mero muestra cu√°nto sube/baja el futuro si cambias la tasa (filas) o el costo de almacenaje (columnas).

üìå **Idea clave**
El futuro replica el spot m√°s los costos netos de financiaci√≥n y almacenamiento.
"""))

# Gr√°ficos de sensibilidad
fig, axs = plt.subplots(1, 3, figsize=(22, 6))

# Sensibilidad al tiempo
tiempos = np.linspace(0.1, 2, 20)
precios = [ValoradorFuturos(futuro_params['precio_spot'],
                           futuro_params['tasa_libre_riesgo'],
                           t,
                           futuro_params['costo_almacenamiento'],
                           futuro_params['rendimiento_conveniencia']).precio_teorico() for t in tiempos]

axs[0].plot(tiempos, precios, linewidth=2.5, color=usach_colors[0])
axs[0].set_title("Evoluci√≥n del Precio Futuro seg√∫n Plazo", fontweight='bold')
axs[0].set_xlabel("Tiempo al Vencimiento (a√±os)")
axs[0].set_ylabel("Precio Futuro ($)")
axs[0].grid(True, alpha=0.3)
axs[0].axhline(y=futuro_params['precio_spot'], color='r', linestyle='--', label='Precio Spot')
axs[0].axvline(x=futuro_params['tiempo'], color='g', linestyle='--', label='Contrato (6 meses)')
axs[0].legend()
axs[0].yaxis.set_major_formatter('${x:,.0f}')

plt.savefig("grafico_sensibilidad.png")

# Convergencia spot-futuro
tiempo_restante = np.linspace(0.5, 0, 100)
precios_convergencia = [futuro_params['precio_spot'] + (futuro_teorico - futuro_params['precio_spot']) * t/0.5 for t in tiempo_restante]

axs[1].plot(tiempo_restante, precios_convergencia, linewidth=2.5, color=usach_colors[1])
axs[1].set_title("Convergencia Precio Futuro-Precio Spot", fontweight='bold')
axs[1].set_xlabel("Tiempo al Vencimiento (a√±os)")
axs[1].set_ylabel("Precio Futuro ($)")
axs[1].invert_xaxis()
axs[1].grid(True, alpha=0.3)
axs[1].axhline(y=futuro_params['precio_spot'], color='r', linestyle='--', label='Precio Spot')
axs[1].legend()
axs[1].yaxis.set_major_formatter('${x:,.0f}')

plt.savefig("grafico_convergencia.png")

# Heatmap de sensibilidad
sns.heatmap(tabla_sensibilidad.pivot(index='Tasa', columns='Costo Almacen', values='Precio Futuro'),
            annot=True, fmt=".2f", cmap="coolwarm", ax=axs[2])
axs[2].set_title("Sensibilidad Precio Futuro (Tasas vs Costos)", fontweight='bold')
axs[2].set_xlabel("Costo de Almacenamiento")
axs[2].set_ylabel("Tasa Libre de Riesgo")

plt.savefig("grafico_heatmap.png")

plt.tight_layout()
plt.show()


# =============================================================================
# 4. VALORACI√ìN DE FORWARDS (MEJORADA)
# =============================================================================
print_header("4. valoraci√≥n de forwards mejorada", level=1)

"""
Teor√≠a:
Los forwards son similares a los futuros pero no estandarizados y transados OTC.
El precio forward se calcula como (Hull, 2018, Cap. 5):

F = S √ó e^{(r + c)T}

Donde:
F = Precio forward
S = Precio spot
r = Tasa libre de riesgo
c = Costo de almacenamiento
T = Tiempo hasta vencimiento

El valor del contrato forward se calcula como:

f = (F_actual - K) √ó e^{-rT}

Donde:
f = Valor del contrato
F_actual = Precio forward actual
K = Precio pactado originalmente
"""

class ValoradorForward:
    def __init__(self, precio_spot, tasa, tiempo, costo_almacenamiento=0):
        """
        Valoraci√≥n de Forward Rate Agreement (FRA)

        Par√°metros:
        precio_spot: Precio spot del activo
        tasa: Tasa libre de riesgo
        tiempo: Tiempo hasta vencimiento (a√±os)
        costo_almacenamiento: Costo de carry (anualizado)
        """
        self.precio_spot = precio_spot
        self.tasa = tasa
        self.tiempo = tiempo
        self.costo_almacenamiento = costo_almacenamiento

    def precio_forward_actual(self):
        """Precio forward actual F0"""
        return self.precio_spot * np.exp((self.tasa + self.costo_almacenamiento) * self.tiempo)

    def valor_contrato(self, precio_pactado):
        """Valor actual del contrato forward (f)"""
        forward_actual = self.precio_forward_actual()
        return (forward_actual - precio_pactado) * np.exp(-self.tasa * self.tiempo)

# Ejemplo: Forward de oro
forward_params = {
    'precio_spot': spot_oro,  # Precio real actualizado
    'tasa': 0.0425,  # Actualizado a tasa actual
    'tiempo': 1,
    'costo_almacenamiento': 0.01  # Costo de almacenamiento a√±adido
}

precio_pactado = 1850.0
valorador_forward = ValoradorForward(**forward_params)
forward_actual = valorador_forward.precio_forward_actual()
valor_contrato = valorador_forward.valor_contrato(precio_pactado)

tabla_forward = pd.DataFrame({
    'Par√°metro': ['Precio Spot (S0)', 'Tasa Libre Riesgo', 'Costo Almacenamiento', 'Tiempo (a√±os)',
                  'Precio Pactado (K)', 'Precio Forward Actual (F0)', 'Valor Contrato (f)'],
    'Valor': [f"${forward_params['precio_spot']:.2f}",
              f"{forward_params['tasa']*100:.2f}%",
              f"{forward_params['costo_almacenamiento']*100:.2f}%",
              f"{forward_params['tiempo']:.1f}",
              f"${precio_pactado:.2f}",
              f"${forward_actual:.2f}",
              f"${valor_contrato:.2f}"]
})

print("\nResultados de valoraci√≥n de forwards:")
print(tabla_forward.to_string(index=False))
print(f"\nInterpretaci√≥n: El contrato est√° {'in-the-money' if valor_contrato > 0 else 'out-of-the-money'} "
      f"para el {'comprador' if valor_contrato > 0 else 'vendedor'}")

# =============================================================================
# MARKDOWN 4: FORWARD ORO - VALOR Y SENSIBILIDADES
# =============================================================================
display(Markdown("""
## 4 | Forward Oro ‚Äì Valor y sensibilidades

### üîç Qu√© mirar

* **Valor vs Precio spot** ‚Äì l√≠nea casi 45¬∞: 1 $ que suba el oro ‚âà 1 $ que sube tu contrato.
* **Valor vs Tiempo** ‚Äì el valor extra se ‚Äúdesinfla‚Äù a medida que pasa el tiempo.

üìå **Idea clave**
El forward es pr√°cticamente un ‚Äúdelta 1‚Äù sobre el spot, pero s√≥lo hasta la fecha pactada.
"""))

# Gr√°ficos de sensibilidad
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

# Sensibilidad al precio spot
precios_spot = np.linspace(1600, 2000, 50)
valores_contrato = [ValoradorForward(p,
                                    forward_params['tasa'],
                                    forward_params['tiempo'],
                                    forward_params['costo_almacenamiento']).valor_contrato(precio_pactado) for p in precios_spot]

axs[0].plot(precios_spot, valores_contrato, linewidth=2.5, color=usach_colors[0])
axs[0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axs[0].axvline(x=forward_params['precio_spot'], color='r', linestyle='--', label='Precio Spot Actual')
axs[0].set_title("Valor del Contrato Forward vs Precio Spot", fontweight='bold')
axs[0].set_xlabel("Precio Spot del Oro ($)")
axs[0].set_ylabel("Valor del Contrato ($)")
axs[0].grid(True, alpha=0.3)
axs[0].legend()

# Sensibilidad al tiempo
tiempos_forward = np.linspace(0.1, 2, 20)
valores_tiempo = [ValoradorForward(forward_params['precio_spot'],
                                    forward_params['tasa'],
                                    t,
                                    forward_params['costo_almacenamiento']).valor_contrato(precio_pactado) for t in tiempos_forward]

axs[1].plot(tiempos_forward, valores_tiempo, linewidth=2.5, color=usach_colors[1])
axs[1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axs[1].axvline(x=forward_params['tiempo'], color='r', linestyle='--', label='Plazo Actual (1 a√±o)')
axs[1].set_title("Valor del Contrato Forward vs Tiempo", fontweight='bold')
axs[1].set_xlabel("Tiempo al Vencimiento (a√±os)")
axs[1].set_ylabel("Valor del Contrato ($)")
axs[1].grid(True, alpha=0.3)
axs[1].legend()

plt.tight_layout()
plt.show()
plt.savefig("grafico_forward.png")


# =============================================================================
# 5. VALORACI√ìN DE OPCIONES (MEJORADA SIN GRIEGAS)
# =============================================================================
print_header("5. valoraci√≥n de opciones mejorada (sin griegas)", level=1)

"""
Teor√≠a:
El valor de una opci√≥n est√° compuesto por:
- Valor intr√≠nseco (VI): Diferencia entre precio subyacente y strike
- Valor del tiempo (VT): Prima por volatilidad y tiempo restante

Para una call:
VI = max(S - K, 0)
VT = Prima total - VI

Factores que afectan el precio:
1. Precio subyacente (S) ‚Üë ‚Üí Call ‚Üë / Put ‚Üì
2. Precio ejercicio (K) ‚Üë ‚Üí Call ‚Üì / Put ‚Üë
3. Tiempo expiraci√≥n ‚Üë ‚Üí Prima ‚Üë
4. Volatilidad ‚Üë ‚Üí Prima ‚Üë
5. Tasas inter√©s ‚Üë ‚Üí Call ‚Üë / Put ‚Üì
6. Dividendos ‚Üë ‚Üí Call ‚Üì / Put ‚Üë
"""

class ValoradorOpciones:
    def __init__(self, S, K, T, r, sigma, tipo='call'):
        self.S = S          # Precio actual del subyacente
        self.K = K          # Precio de ejercicio
        self.T = T          # Tiempo al vencimiento (a√±os)
        self.r = r          # Tasa libre de riesgo anual
        self.sigma = sigma  # Volatilidad anualizada
        self.tipo = tipo    # 'call' o 'put'

    def precio_bs(self):
        """Calcula precio con Black-Scholes"""
        # Manejar casos l√≠mite
        if self.T <= 0:
            # Al vencimiento, devolver valor intr√≠nseco
            return self.valor_intrinseco()

        d1 = (np.log(self.S/self.K) + (self.r + self.sigma**2/2)*self.T) / (self.sigma*np.sqrt(self.T))
        d2 = d1 - self.sigma*np.sqrt(self.T)

        if self.tipo == 'call':
            return self.S*norm.cdf(d1) - self.K*np.exp(-self.r*self.T)*norm.cdf(d2)
        else:
            return self.K*np.exp(-self.r*self.T)*norm.cdf(-d2) - self.S*norm.cdf(-d1)

    def valor_intrinseco(self):
        """Calcula el valor intr√≠nseco"""
        if self.tipo == 'call':
            return max(self.S - self.K, 0)
        else:
            return max(self.K - self.S, 0)

    def valor_tiempo(self):
        """Calcula el valor del tiempo"""
        return self.precio_bs() - self.valor_intrinseco()

    def estado_dinero(self):
        """Determina si la opci√≥n est√° in/at/out of the money"""
        if self.tipo == 'call':
            if self.S > self.K: return "In the Money (ITM)"
            elif self.S < self.K: return "Out of the Money (OTM)"
            else: return "At the Money (ATM)"
        else:
            if self.K > self.S: return "In the Money (ITM)"
            elif self.K < self.S: return "Out of the Money (OTM)"
            else: return "At the Money (ATM)"

# Ejemplo: Opci√≥n call
opcion_params = {
    'S': 150.0,
    'K': 155.0,
    'T': 0.5,
    'r': 0.0425,
    'sigma': 0.25,
    'tipo': 'call'
}

opcion_call = ValoradorOpciones(**opcion_params)
precio_call = opcion_call.precio_bs()
vi_call = opcion_call.valor_intrinseco()
vt_call = opcion_call.valor_tiempo()
estado_call = opcion_call.estado_dinero()

# Ejemplo para put
opcion_put = ValoradorOpciones(opcion_params['S'], opcion_params['K'], opcion_params['T'],
                              opcion_params['r'], opcion_params['sigma'], 'put')
precio_put = opcion_put.precio_bs()
vi_put = opcion_put.valor_intrinseco()
vt_put = opcion_put.valor_tiempo()
estado_put = opcion_put.estado_dinero()

# Validaci√≥n de paridad put‚Äìcall
left_side = precio_call + opcion_params['K'] * np.exp(-opcion_params['r'] * opcion_params['T'])
right_side = precio_put + opcion_params['S']
diferencia = abs(left_side - right_side)
if diferencia < 1e-8:
    print("‚Äì Paridad put‚Äìcall verificada.")
else:
    print(f"Advertencia: Paridad put‚Äìcall no se cumple. Diferencia = {diferencia:.10f}")

# Tablas de resultados
tabla_opciones = pd.DataFrame({
    'Par√°metro': ['Precio Subyacente', 'Precio Ejercicio', 'Tiempo Vencimiento (a√±os)',
                  'Tasa Libre Riesgo', 'Volatilidad', 'Tipo Opci√≥n'],
    'Valor': [f"${opcion_params['S']:.2f}",
              f"${opcion_params['K']:.2f}",
              f"{opcion_params['T']:.2f}",
              f"{opcion_params['r']*100:.2f}%",
              f"{opcion_params['sigma']*100:.2f}%",
              "Call"]
})

tabla_valores_call = pd.DataFrame({
    'Componente': ['Precio Total', 'Valor Intr√≠nseco', 'Valor Tiempo', 'Estado'],
    'Valor': [f"${precio_call:.2f}", f"${vi_call:.2f}", f"${vt_call:.2f}", estado_call]
})

tabla_valores_put = pd.DataFrame({
    'Componente': ['Precio Total', 'Valor Intr√≠nseco', 'Valor Tiempo', 'Estado'],
    'Valor': [f"${precio_put:.2f}", f"${vi_put:.2f}", f"${vt_put:.2f}", estado_put]
})

print("\nPar√°metros de la opci√≥n:")
print(tabla_opciones.to_string(index=False))

print("\nValoraci√≥n Call:")
print(tabla_valores_call.to_string(index=False))

print("\nValoraci√≥n Put:")
print(tabla_valores_put.to_string(index=False))

# =============================================================================
# GR√ÅFICOS DE COMPONENTES Y SENSIBILIDADES CON AN√ÅLISIS EXPLICATIVO
# =============================================================================
print_header("an√°lisis visual de componentes", level=2)

# --- Bloque de an√°lisis autom√°tico -------------------------------------------------
print("\n" + "="*80)
print("üîç AN√ÅLISIS AUTOM√ÅTICO DE GR√ÅFICOS")
print("="*80)

print("\nüîπ Gr√°fico 1: Componentes vs S")
print("  ‚Äì Intr√≠nseco=0 hasta S=K; pendiente=1 luego.")
print("  ‚Äì VT m√°ximo en ATM; VT‚Üí0 en OTM profundo e ITM profundo.")
print("  ‚Äì L√≠nea roja = suma exacta intr√≠nseco+VT (se verifica por assert).")
print("  ‚Äì L√≠nea punteada en S=K marca el paso OTM‚ÜíITM.")


# Gr√°fico 1: Componentes de valor para diferentes precios subyacentes
precios_subyacente = np.linspace(100, 200, 100)
vi_values = [max(p - opcion_params['K'], 0) for p in precios_subyacente]
vt_values = []
precio_total_values = []

for p in precios_subyacente:
    valorador = ValoradorOpciones(p, opcion_params['K'], opcion_params['T'],
                                 opcion_params['r'], opcion_params['sigma'])
    vt = valorador.valor_tiempo()
    vt_values.append(vt)
    precio_total_values.append(valorador.precio_bs())

# Chequeo de suma intr√≠nseco+tiempo = precio
total_calculado = [vi + vt for vi, vt in zip(vi_values, vt_values)]
if np.allclose(total_calculado, precio_total_values, atol=1e-6):
    print("‚Äì Descomposici√≥n intr√≠nseco+tiempo OK en toda la grilla S.")
else:
    print("Advertencia: Descomposici√≥n intr√≠nseco+tiempo no coincide")

plt.figure(figsize=(14, 8))
plt.plot(precios_subyacente, vi_values, 'b-', linewidth=2.5, label='Valor Intr√≠nseco')
plt.plot(precios_subyacente, vt_values, 'g-', linewidth=2.5, label='Valor Tiempo')
plt.plot(precios_subyacente, precio_total_values, 'r--', linewidth=2.5, label='Precio Total')
plt.axvline(x=opcion_params['K'], color='k', linestyle='--', alpha=0.7, label='Strike Price')

# Anotaci√≥n mejorada para punto ATM
atm_index = np.abs(precios_subyacente - opcion_params['K']).argmin()
plt.annotate(f"ATM: M√°ximo VT = ${vt_values[atm_index]:.2f}",
            xy=(opcion_params['K'], vt_values[atm_index]),
            xytext=(opcion_params['K']+10, vt_values[atm_index]),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

plt.title("Componentes del Valor de una Call", fontweight='bold', fontsize=14)
plt.xlabel("Precio del Subyacente (S)", fontsize=12)
plt.ylabel("Valor (USD)", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11, loc='upper left')
plt.tight_layout()
plt.savefig("componentes_opcion.png")
plt.show()

# --- An√°lisis para Gr√°fico 2 ---
print("\nüîπ Gr√°fico 2: VT vs T")
print("  ‚Äì Curva c√≥ncava: theta m√°s negativo para T muy peque√±o.")
print("  ‚Äì VT‚Üí0 cuando T‚Üí0 (precio‚Üíintr√≠nseco).")
print("  ‚Äì Mon√≥tono creciente con T.")
tiempos = np.linspace(0.001, 1.0, 100)  # M√°s puntos cerca de cero
valores_tiempo = [ValoradorOpciones(opcion_params['S'], opcion_params['K'], t,
                  opcion_params['r'], opcion_params['sigma']).valor_tiempo()
                 for t in tiempos]

plt.figure(figsize=(14, 8))
plt.plot(tiempos, valores_tiempo, 'm-', linewidth=2.5)

# Resaltar comportamiento cerca de T=0
plt.axvline(x=0.1, color='gray', linestyle='--', alpha=0.5)
plt.annotate("Decaimiento acelerado\n(theta m√°s negativo)",
            xy=(0.05, valores_tiempo[10]),
            xytext=(0.3, valores_tiempo[10]),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

# Marcar l√≠mite cuando T‚Üí0
plt.annotate("VT ‚Üí 0 cuando T ‚Üí 0",
            xy=(0.01, valores_tiempo[0]),
            xytext=(0.15, valores_tiempo[0]/2),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

plt.title("Evoluci√≥n del Valor del Tiempo", fontweight='bold', fontsize=14)
plt.xlabel("Tiempo al Vencimiento (a√±os)", fontsize=12)
plt.ylabel("Valor del Tiempo (USD)", fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("valor_tiempo.png")
plt.show()

# --- An√°lisis para Gr√°fico 3 ---
# Calcular œÉ* donde d1=0 (m√°xima vega)
S = opcion_params['S']
K = opcion_params['K']
T = opcion_params['T']
r = opcion_params['r']

def d1(sigma):
    return (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))

# Resolver d1(sigma) = 0
sigma_star = fsolve(d1, 0.2)[0]  # Valor inicial 0.2

print(f"\nüîπ Gr√°fico 3: VT vs œÉ")
print(f"  ‚Äì VT crece con œÉ, pero se satura en S cuando œÉ‚Üí‚àû.")
print(f"  ‚Äì Vega m√°xima en œÉ* (d1=0) ‚âà {sigma_star:.2f}.")
print("  ‚Äì Para œÉ‚Üí0, VT‚Üí0; chequeo de l√≠mites extremos.")

# Gr√°fico 3: Sensibilidad a la volatilidad
volatilidades = np.linspace(0.01, 1.5, 100)  # Rango ampliado para mostrar saturaci√≥n
valores_vol = [ValoradorOpciones(opcion_params['S'], opcion_params['K'], opcion_params['T'],
                                opcion_params['r'], v).valor_tiempo()
              for v in volatilidades]

plt.figure(figsize=(14, 8))
plt.plot(volatilidades, valores_vol, 'c-', linewidth=2.5)

# Anotaciones para comportamiento en extremos
plt.annotate("œÉ‚Üí0: VT‚Üí0",
            xy=(0.05, valores_vol[5]),
            xytext=(0.1, valores_vol[5] + 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

plt.annotate(f"VT‚Üí${S:.2f} cuando œÉ‚Üí‚àû",
            xy=(1.2, S-2),
            xytext=(0.8, S-5),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

# L√≠nea de saturaci√≥n en S
plt.axhline(y=S, color='r', linestyle='--', alpha=0.5, label='L√≠mite S')

# Marcar punto de m√°xima vega
max_vega_idx = np.argmin(np.abs(volatilidades - sigma_star))
plt.axvline(x=sigma_star, color='gray', linestyle='--', alpha=0.7)
plt.annotate(f"œÉ*‚âà{sigma_star:.2f} (m√°xima vega)",
            xy=(sigma_star, valores_vol[max_vega_idx]),
            xytext=(sigma_star+0.1, valores_vol[max_vega_idx]-2),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

plt.title("Sensibilidad del Valor del Tiempo a la Volatilidad", fontweight='bold', fontsize=14)
plt.xlabel("Volatilidad œÉ (en decimales)", fontsize=12)
plt.ylabel("Valor Tiempo (USD)", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig("volatilidad_tiempo.png")
plt.show()

# --- Resumen clave ---
print("\n" + "="*80)
print("üìå RESUMEN CLAVE Y FACTORES FUNDAMENTALES")
print("="*80)
print("Descomposici√≥n de la prima")
print("")
print("ùê∂ = max(ùëÜ‚àíùêæ,0) (VI) + ùëâùëá (m√°x. en ATM)")
print("")
print("Factores que afectan ùê∂:")
print("")
print("Precio spot ùëÜ: ‚Üë ‚áí ùê∂ ‚Üë")
print("Strike ùêæ: ‚Üë ‚áí ùê∂ ‚Üì")
print("Tiempo ùëá: ‚Üë ‚áí ùê∂ ‚Üë")
print("Volatilidad ùúé: ‚Üë ‚áí ùê∂ ‚Üë (pendiente m√°xima en œÉ*)")
print("Tasa ùëü: ‚Üë ‚áí ùê∂ ‚Üë")
print("Dividendos ùëë: ‚Üë ‚áí ùê∂ ‚Üì")
print("")
print("L√≠mites y paridad")
print("")
print("Cotas (sin dividendos):")
print("   max(ùëÜ‚àíùêæùëí‚àíùëüùëá,‚Äâ0)‚ÄÖ‚Ää‚â§‚ÄÖ‚Ääùê∂‚ÄÖ‚Ää‚â§‚ÄÖ‚ÄäùëÜ")
print("")
print("Put‚ÄìCall Parity (europea):")
print("   ùê∂+ùêæùëí‚àíùëüùëá=ùëÉ+ùëÜ")
print("")
print("Comportamiento en extremos")
print("")
print("ùëá‚Üí0 o ùúé‚Üí0: ùëâùëá‚Üí0.")
print("")
print("ùúé‚Üí‚àû (para OTM puro): ùê∂‚ÜíùëÜ‚ÄÖ‚Ää‚áí‚ÄÖ‚Ääùëâùëá‚ÜíùëÜ.")
print("")
print("Ejercicio temprano")
print("")
print("Call americano: nunca √≥ptimo antes de T.")
print("Put americano: puede valer m√°s que su equivalente europeo por ventaja de ejercicio anticipado.")
print("="*80)

# =============================================================================
# 6. VALORACI√ìN DE SWAPS (MEJORADA)
# =============================================================================
print_header("6. valoraci√≥n de swaps mejorada", level=1)

"""
Teor√≠a:
Un swap de tasas de inter√©s puede verse como una cartera de dos bonos:
1. Un bono de tasa fija (pata fija)
2. Un bono de tasa flotante (pata flotante)

El valor del swap es la diferencia entre estos dos bonos (Hull, 2018, Cap. 7):

V_swap = B_flotante - B_fijo

Para la valoraci√≥n:
1. Construir curva de tasas cero mediante bootstrapping
2. Calcular factores de descuento
3. Valorar cada pata como el VP de sus flujos futuros
"""

class ValoradorSwap:
    def __init__(self, notional, curva_fija, curva_flotante, tiempos, verbose=True):
        """
        Valoraci√≥n de swap usando curva de descuento

        Par√°metros:
        notional: Principal nocional
        curva_fija: Lista de tasas fijas para cada per√≠odo
        curva_flotante: Lista de tasas flotantes observadas
        tiempos: Lista de tiempos hasta cada pago (a√±os)
        verbose: Controla la impresi√≥n de resultados intermedios
        """
        self.notional = notional
        self.curva_fija = curva_fija
        self.curva_flotante = curva_flotante
        self.tiempos = tiempos
        self.verbose = verbose

        # =====================================================================
        # 6.1 Bootstrapping de curva cero y factores de descuento
        # =====================================================================
        if verbose:
            print_header("6.1 bootstrapping de curva cero", level=2)

        # Construir factores de descuento mediante bootstrapping
        self.factores_descuento = []

        # Para el primer per√≠odo: DF = 1 / (1 + r*T)
        t0 = tiempos[0]
        r0 = curva_flotante[0]
        df0 = 1 / (1 + r0 * t0)
        self.factores_descuento.append(df0)

        # Para per√≠odos subsecuentes
        for i in range(1, len(tiempos)):
            ti = tiempos[i]
            ri = curva_flotante[i]
            # Suponiendo pagos intermedios
            dfi = 1 / (1 + ri * (ti - tiempos[i-1]))
            self.factores_descuento.append(dfi)

        if verbose:
            print("\nFactores de descuento calculados:")
            tabla_descuento = pd.DataFrame({
                'Per√≠odo': [f"t={t:.1f}" for t in tiempos],
                'Tasa Flotante': [f"{r*100:.4f}%" for r in curva_flotante],
                'Factor Descuento': [f"{df:.6f}" for df in self.factores_descuento]
            })
            print(tabla_descuento.to_string(index=False))

    def valor_pata_fija(self):
        """Calcula el valor presente de la pata fija"""
        vp_fijo = 0
        for i in range(len(self.tiempos)):
            periodo = self.tiempos[i] - (self.tiempos[i-1] if i>0 else 0)
            flujo = self.notional * self.curva_fija[i] * periodo
            vp_fijo += flujo * self.factores_descuento[i]
        return vp_fijo

    def valor_pata_flotante(self):
        """Calcula el valor presente de la pata flotante"""
        # Para swap flotante, el pr√≥ximo pago es conocido
        # y el valor despu√©s del pr√≥ximo pago es el nominal
        if len(self.tiempos) > 0:
            periodo = self.tiempos[0]
            flujo = self.notional * self.curva_flotante[0] * periodo
            vp_flotante = flujo * self.factores_descuento[0]
            vp_flotante += self.notional * self.factores_descuento[-1]  # Principal al final
            return vp_flotante
        return 0

    def valor_swap(self):
        """Valor del swap para el pagador fijo"""
        return self.valor_pata_flotante() - self.valor_pata_fija()

    def sensibilidad_bp(self, shift=0.0001):
        """Calcula sensibilidad a un movimiento de 1bp en la curva"""
        valor_original = self.valor_swap()

        # Aplicar shift a la curva flotante
        curva_shifted = [r + shift for r in self.curva_flotante]
        swap_shifted = ValoradorSwap(self.notional, self.curva_fija, curva_shifted, self.tiempos, verbose=False)
        valor_shifted = swap_shifted.valor_swap()

        return (valor_shifted - valor_original)  # DV01 (d√≥lar por 1bp)

# Ejemplo: Swap de 2 a√±os con curva bootstrapped
notional = 1_000_000
curva_fija = [0.04, 0.04]  # Tasas fijas
curva_flotante = [0.035, 0.036]  # Tasas flotantes observadas
tiempos = [1.0, 2.0]

swap = ValoradorSwap(notional, curva_fija, curva_flotante, tiempos)
valor_swap_result = swap.valor_swap()
sensibilidad_1bp = swap.sensibilidad_bp()

tabla_swap = pd.DataFrame({
    'Per√≠odo': ['A√±o 1', 'A√±o 2'],
    'Tiempo': tiempos,
    'Tasa Fija': [f"{curva_fija[0]*100:.4f}%", f"{curva_fija[1]*100:.4f}%"],
    'Tasa Flotante': [f"{curva_flotante[0]*100:.4f}%", f"{curva_flotante[1]*100:.4f}%"],
    'Factor Descuento': [f"{swap.factores_descuento[0]:.6f}",
                         f"{swap.factores_descuento[1]:.6f}"]
})

print("\nResultados de valoraci√≥n:")
print(f"Valor del swap: ${valor_swap_result:,.2f}")
print(f"Sensibilidad a 1bp: ${sensibilidad_1bp:,.2f} (DV01)")
print(tabla_swap.to_string(index=False))

# =============================================================================
# MARKDOWN 6: SWAP IRS - VALOR, DV01 Y CURVA CERO
# =============================================================================
display(Markdown("""
## 6 | Swap IRS ‚Äì Valor, DV01 y curva cero

### üîç Qu√© mirar

* **Valor vs movimiento de curva** ‚Äì l√≠nea casi recta: poco efecto de convexidad. La pendiente es el **DV01** (~ \$7,6 por bp por \$1 MM).
* **Curva cero** ‚Äì tasas bootstrapped: 3,50 % a 3,60 %. Confirma que los inputs son consistentes.

üìå **Idea clave**
Un swap es un instrumento ‚Äúpuramente de duraci√≥n‚Äù para gestionar brechas de tasa.
"""))

# Gr√°fico de sensibilidad
shifts = np.linspace(-0.02, 0.02, 50)
valores_swap = []

for shift in shifts:
    curva_shifted = [r + shift for r in curva_flotante]
    swap_shifted = ValoradorSwap(notional, curva_fija, curva_shifted, tiempos, verbose=False)
    valores_swap.append(swap_shifted.valor_swap())

plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(shifts*10000, valores_swap, linewidth=2.5, color=usach_colors[0])
plt.axvline(x=0, color='r', linestyle='--', alpha=0.7)
plt.title("Sensibilidad del Swap a Movimientos de la Curva", fontweight='bold')
plt.xlabel("Cambio en la Curva (bps)")
plt.ylabel("Valor del Swap ($)")
plt.grid(True, alpha=0.3)
plt.annotate(f"Valor actual: ${valor_swap_result:,.2f}\nDV01: ${sensibilidad_1bp:,.2f}",
            xy=(-150, valor_swap_result*1.5),
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

# Curva cero
plt.subplot(1, 2, 2)
plt.plot(tiempos, [r*100 for r in curva_flotante], 'o-', linewidth=2.5, color=usach_colors[1])
plt.title("Curva Cero Bootstrapped", fontweight='bold')
plt.xlabel("Tiempo (a√±os)")
plt.ylabel("Tasa (%)")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
plt.savefig("grafico_swap.png")

# =============================================================================
# 7. VALORACI√ìN DE FRAs (MEJORADA)
# =============================================================================
print_header("7. valoraci√≥n de fras mejorada", level=1)

"""
Teor√≠a:
Un Forward Rate Agreement (FRA) es un contrato OTC que fija una tasa de inter√©s
para un per√≠odo futuro. Su valor se calcula como (Hull, 2018, Cap. 7):

V = N √ó (R_fra - R_mercado) √ó T / (1 + R_mercado √ó T)

Donde:
N = Principal nocional
R_fra = Tasa pactada en el FRA
R_mercado = Tasa de mercado actual para el per√≠odo
T = Duraci√≥n del per√≠odo cubierto
"""

class ValoradorFRA:
    def __init__(self, notional, tasa_fra, tasa_mercado, tiempo):
        """
        Valoraci√≥n de Forward Rate Agreement (FRA)

        Par√°metros:
        notional: Principal nocional
        tasa_fra: Tasa pactada en el FRA
        tasa_mercado: Tasa de mercado actual para el per√≠odo
        tiempo: Per√≠odo cubierto por el FRA (a√±os)
        """
        self.notional = notional
        self.tasa_fra = tasa_fra
        self.tasa_mercado = tasa_mercado
        self.tiempo = tiempo

    def valor(self):
        """Valor presente del FRA"""
        return self.notional * (self.tasa_fra - self.tasa_mercado) * self.tiempo / (1 + self.tasa_mercado * self.tiempo)

    def sensibilidad_bp(self, shift=0.0001):
        """Sensibilidad a 1bp en la tasa de mercado"""
        valor_original = self.valor()
        fra_shifted = ValoradorFRA(self.notional, self.tasa_fra, self.tasa_mercado + shift, self.tiempo)
        return fra_shifted.valor() - valor_original

# Ejemplo: FRA 6x12
fra_params = {
    'notional': 1_000_000,
    'tasa_fra': 0.05,
    'tasa_mercado': 0.048,
    'tiempo': 0.5
}

fra = ValoradorFRA(**fra_params)
valor_fra = fra.valor()
sensibilidad_1bp_fra = fra.sensibilidad_bp()

tabla_fra = pd.DataFrame({
    'Par√°metro': ['Principal Nocional', 'Tasa FRA', 'Tasa Mercado', 'Per√≠odo (a√±os)', 'Valor FRA'],
    'Valor': [f"${fra_params['notional']:,.0f}",
              f"{fra_params['tasa_fra']*100:.4f}%",
              f"{fra_params['tasa_mercado']*100:.4f}%",
              f"{fra_params['tiempo']:.4f}",
              f"${valor_fra:,.2f}"]
})

print("\nResultados de valoraci√≥n de FRA:")
print(tabla_fra.to_string(index=False))
print(f"Sensibilidad a 1bp: ${sensibilidad_1bp_fra:,.2f} (DV01)")

# Relaci√≥n con swaps
print("\nRelaci√≥n entre FRA y Swaps:")
print("Un swap puede interpretarse como una cartera de FRAs consecutivos")

# =============================================================================
# MARKDOWN 7: FRA 6√ó12 - VALOR Y DV01
# =============================================================================
display(Markdown("""
## 7 | FRA 6√ó12 ‚Äì Valor y DV01

### üîç Qu√© mirar

* **Valor vs bps** ‚Äì relaci√≥n perfectamente lineal (convexidad ‚âà 0).
* **DV01 ‚Äì\$48,9** ‚Üí por cada bp de subida pierdes \$48,9 por mill√≥n de nocional.

üìå **Idea clave**
√ötil para cubrir un tramo concreto de la curva sin compromisos a m√°s largo plazo.
"""))

# Gr√°fico de sensibilidad
shifts_fra = np.linspace(-0.02, 0.02, 50)
valores_fra = [ValoradorFRA(fra_params['notional'], fra_params['tasa_fra'],
               fra_params['tasa_mercado'] + s, fra_params['tiempo']).valor() for s in shifts_fra]

plt.figure(figsize=(10, 6))
plt.plot(shifts_fra*10000, valores_fra, linewidth=2.5, color=usach_colors[1])
plt.axvline(x=0, color='r', linestyle='--', alpha=0.7)
plt.title("Sensibilidad del FRA a Movimientos de Tasas", fontweight='bold')
plt.xlabel("Cambio en Tasa de Mercado (bps)")
plt.ylabel("Valor del FRA ($)")
plt.grid(True, alpha=0.3)
plt.annotate(f"Valor actual: ${valor_fra:,.2f}\nDV01: ${sensibilidad_1bp_fra:,.2f}",
            xy=(-150, valor_fra*1.5),
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
plt.tight_layout()
plt.show()
plt.savefig("grafico_fra.png")

# =============================================================================
# RESUMEN GENERAL
# =============================================================================
print_header("resumen general de valoraciones", level=1)

# Manejo seguro de variables que podr√≠an no estar definidas
try:
    retorno_anual
    volatilidad_anual
    valor_ddm
    beta
except NameError:
    retorno_anual = 0
    volatilidad_anual = 0
    valor_ddm = 0
    beta = np.nan

resumen_general = pd.DataFrame({
    'Activo': ['Bono', 'Acci√≥n (MSFT)', 'Futuro Petr√≥leo', 'Forward Oro', 'Opci√≥n Call', 'Swap', 'FRA'],
    'Valor': [f"${precio_bono:.2f}",
              f"${valor_ddm:.2f}" if 'valor_ddm' in locals() else "N/A",
              f"${futuro_teorico:.2f}",
              f"${valor_contrato:.2f}",
              f"${precio_call:.2f}",
              f"${valor_swap_result:,.2f}",
              f"${valor_fra:,.2f}"],
    'Rentabilidad': [f"{bono_params['tasa_cupon']*100:.1f}% (cup√≥n)",
                     f"{retorno_anual*100:.2f}%" if 'retorno_anual' in locals() else "N/A",
                     'N/A',
                     'N/A',
                     'Variable (apalanc.)',
                     'Spread 0.4%',
                     'Diferencial 0.2%'],
    'Riesgo': [f'Duraci√≥n {duracion:.2f} a√±os',
               f'Volat. {volatilidad_anual*100:.2f}%' if 'volatilidad_anual' in locals() else "N/A",
               'Precio spot',
               'Contraparte',
               'Alto (apalanc.)',
               'Tasa inter√©s',
               'Tasa mercado']
})

print(resumen_general.to_string(index=False))
print("\n" + "="*80)

# =============================================================================
# CONCLUSI√ìN COMPARATIVA
# =============================================================================
display(Markdown("""
## 8 | Comparaci√≥n final ‚Äì ¬øQu√© activo rinde mejor por unidad de riesgo?

| Activo | M√©trica clave | Resultado | Tipo de inversor |
|--------|---------------|-----------|------------------|
| **MSFT** | **Sharpe 0,94** (retorno/vol) | M√°s eficiente | Largo plazo con tolerancia a mercado accionario |
| **Bono 5 a 6 %** | Duraci√≥n 4,38 / Convexidad 23,5 | Protecci√≥n moderada ante ca√≠das de tasa (ingresa convexidad). | Conveniente para inversor conservador que prioriza estabilidad. |
| **Swap IRS** | **DV01 \$7,6** por \$1 M | Bajo riesgo por bp ‚Üí fine-tuning de descalce ALM. | √ötil a bancos y tesorer√≠as que gestionan gap de tasa. |
| **Opci√≥n Call** | **Vega 0,42 \$/%** | Mayor exposici√≥n a cambios en œÉ. | Preferida por traders de volatilidad o coberturas de earning surprise. |

üí° **Conclusi√≥n:** no existe el activo universalmente superior. La selecci√≥n √≥ptima depende del perfil de riesgo y horizonte temporal del inversor:

- **Conservador-ingreso fijo** ‚Üí Bono / Swap corto
- **Balance bancario** ‚Üí Swap largo DV01 bajo
- **Alto-crecimiento & tolerancia riesgo** ‚Üí MSFT (Sharpe elevado)
- **Apostar a volatilidad** ‚Üí Opci√≥n Call con vega dominante

> **Principio fundamental:**
> Maximizar retorno esperado por unidad de riesgo relevante
> (volatilidad, DV01, Vega)
> sujeto a las restricciones del inversor.
"""))

# Bibliograf√≠a
print_header("referencias bibliogr√°ficas", level=1)
print("1. Hull, J. C. (2018). Options, Futures and Other Derivatives (10th ed.). Pearson.")
print("2. Gitman, L. J., & Joehnk, M. D. (2015). Fundamentals of Investing (12th ed.). Pearson.")
print("3. Damodaran, A. (2012). Investment Valuation: Tools and Techniques for Determining the Value of Any Asset (3rd ed.). Wiley.")
print("4. Tuckman, B., & Serrat, A. (2021). Fixed Income Securities: Tools for Today's Markets (4th ed.). Wiley.")
print("5. U.S. Department of the Treasury. (2025). Daily Treasury Yield Curve Rates.")
```

```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
import yfinance as yf
!pip install fpdf
from scipy.stats import norm
from datetime import datetime, date
from scipy.optimize import fsolve
from fpdf import FPDF
import matplotlib.image as mpimg
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
import io
import os
import traceback
import urllib.request
import hashlib
import locale
import subprocess

# Configuraci√≥n inicial
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning)
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("colorblind")
pd.options.display.float_format = '{:,.4f}'.format
plt.rcParams.update({
    'figure.figsize': (12, 6),
    'font.size': 12,
    'axes.titlesize': 16,
    'axes.titleweight': 'bold',
    'axes.labelsize': 14,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'grid.alpha': 0.3,
    'savefig.dpi': 300,
    'savefig.bbox': 'tight'
})

# Paleta de colores corporativa USACH
usach_colors = ['#0D3B66', '#1D7A8C', '#A5C882', '#E56B6F', '#9B5DE5']
sns.set_palette(usach_colors)

# Configurar locale para caracteres especiales - Versi√≥n robusta
try:
    locale.setlocale(locale.LC_ALL, 'es_ES.UTF-8')
except:
    try:
        locale.setlocale(locale.LC_ALL, 'es_ES.utf8')
    except:
        try:
            locale.setlocale(locale.LC_ALL, '')
        except:
            print("Advertencia: No se pudo configurar el locale. Usando configuraci√≥n por defecto.")
            locale.setlocale(locale.LC_ALL, 'C')

# =============================================================================
# CLASES PARA VALORACI√ìN
# =============================================================================

class ValoradorBono:
    def __init__(self, valor_nominal, tasa_cupon, a√±os, rendimiento_requerido, frecuencia=2):
        self.valor_nominal = valor_nominal
        self.tasa_cupon = tasa_cupon
        self.a√±os = a√±os
        self.rendimiento_requerido = rendimiento_requerido
        self.frecuencia = frecuencia
        self.n = a√±os * frecuencia
        self.cupon = (tasa_cupon * valor_nominal) / frecuencia
        self.tasa_periodo = rendimiento_requerido / frecuencia

    def precio(self):
        vp_cupones = self.cupon * (1 - (1 + self.tasa_periodo)**-self.n) / self.tasa_periodo
        vp_principal = self.valor_nominal / (1 + self.tasa_periodo)**self.n
        return vp_cupones + vp_principal

    def duracion_macaulay(self):
        precio = self.precio()
        flujos = [self.cupon] * self.n
        flujos[-1] += self.valor_nominal
        t = np.array([(i+1)/self.frecuencia for i in range(self.n)])
        vp_flujos = np.array([cf / (1 + self.tasa_periodo)**(i+1) for i, cf in enumerate(flujos)])
        return np.sum(t * vp_flujos) / precio

    def convexidad(self):
        precio = self.precio()
        flujos = [self.cupon] * self.n
        flujos[-1] += self.valor_nominal
        t = np.array([(i+1)/self.frecuencia for i in range(self.n)])
        convexidad = 0
        for i in range(self.n):
            factor_descuento = 1 / (1 + self.tasa_periodo)**(i+1)
            periodo = t[i]
            convexidad += flujos[i] * periodo * (periodo + 1) * factor_descuento
        return convexidad / (precio * (1 + self.tasa_periodo)**2)

    def ytm_from_price(self, precio_mercado):
        ytm = self.rendimiento_requerido
        max_iter = 100
        tolerancia = 1e-6
        for i in range(max_iter):
            self.rendimiento_requerido = ytm
            self.tasa_periodo = ytm / self.frecuencia
            precio_calculado = self.precio()
            derivada = -self.duracion_macaulay() * precio_calculado / (1 + ytm/self.frecuencia)
            error = precio_calculado - precio_mercado
            if abs(error) < tolerancia:
                break
            ytm -= error / derivada
        return ytm

class ValoradorFuturos:
    def __init__(self, precio_spot, tasa_libre_riesgo, tiempo, costo_almacenamiento=0, rendimiento_conveniencia=0):
        self.precio_spot = precio_spot
        self.tasa_libre_riesgo = tasa_libre_riesgo
        self.tiempo = tiempo
        self.costo_almacenamiento = costo_almacenamiento
        self.rendimiento_conveniencia = rendimiento_conveniencia

    def precio_teorico(self):
        return self.precio_spot * np.exp(
            (self.tasa_libre_riesgo + self.costo_almacenamiento - self.rendimiento_conveniencia) * self.tiempo
        )

class ValoradorForward:
    def __init__(self, precio_spot, tasa, tiempo, costo_almacenamiento=0):
        self.precio_spot = precio_spot
        self.tasa = tasa
        self.tiempo = tiempo
        self.costo_almacenamiento = costo_almacenamiento

    def precio_forward_actual(self):
        return self.precio_spot * np.exp((self.tasa + self.costo_almacenamiento) * self.tiempo)

    def valor_contrato(self, precio_pactado):
        forward_actual = self.precio_forward_actual()
        return (forward_actual - precio_pactado) * np.exp(-self.tasa * self.tiempo)

class ValoradorOpciones:
    def __init__(self, S, K, T, r, sigma, tipo='call'):
        self.S = S
        self.K = K
        self.T = T
        self.r = r
        self.sigma = sigma
        self.tipo = tipo

    def precio_bs(self):
        if self.T <= 0:
            return self.valor_intrinseco()
        d1 = (np.log(self.S/self.K) + (self.r + self.sigma**2/2)*self.T) / (self.sigma*np.sqrt(self.T))
        d2 = d1 - self.sigma*np.sqrt(self.T)
        if self.tipo == 'call':
            return self.S*norm.cdf(d1) - self.K*np.exp(-self.r*self.T)*norm.cdf(d2)
        else:
            return self.K*np.exp(-self.r*self.T)*norm.cdf(-d2) - self.S*norm.cdf(-d1)

    def valor_intrinseco(self):
        if self.tipo == 'call':
            return max(self.S - self.K, 0)
        else:
            return max(self.K - self.S, 0)

    def valor_tiempo(self):
        return self.precio_bs() - self.valor_intrinseco()

class ValoradorSwap:
    def __init__(self, notional, curva_fija, curva_flotante, tiempos):
        self.notional = notional
        self.curva_fija = curva_fija
        self.curva_flotante = curva_flotante
        self.tiempos = tiempos
        self.factores_descuento = []
        t0 = tiempos[0]
        r0 = curva_flotante[0]
        df0 = 1 / (1 + r0 * t0)
        self.factores_descuento.append(df0)
        for i in range(1, len(tiempos)):
            ti = tiempos[i]
            ri = curva_flotante[i]
            dfi = 1 / (1 + ri * (ti - tiempos[i-1]))
            self.factores_descuento.append(dfi)

    def valor_pata_fija(self):
        vp_fijo = 0
        for i in range(len(self.tiempos)):
            periodo = self.tiempos[i] - (self.tiempos[i-1] if i>0 else 0)
            flujo = self.notional * self.curva_fija[i] * periodo
            vp_fijo += flujo * self.factores_descuento[i]
        return vp_fijo

    def valor_pata_flotante(self):
        if len(self.tiempos) > 0:
            periodo = self.tiempos[0]
            flujo = self.notional * self.curva_flotante[0] * periodo
            vp_flotante = flujo * self.factores_descuento[0]
            vp_flotante += self.notional * self.factores_descuento[-1]
            return vp_flotante
        return 0

    def valor_swap(self):
        return self.valor_pata_flotante() - self.valor_pata_fija()

class ValoradorFRA:
    def __init__(self, notional, tasa_fra, tasa_mercado, tiempo):
        self.notional = notional
        self.tasa_fra = tasa_fra
        self.tasa_mercado = tasa_mercado
        self.tiempo = tiempo

    def valor(self):
        return self.notional * (self.tasa_fra - self.tasa_mercado) * self.tiempo / (1 + self.tasa_mercado * self.tiempo)

# =============================================================================
# CLASE PARA GENERAR PDF (MEJORADA)
# =============================================================================

class PDFReport(FPDF):
    def __init__(self):
        super().__init__(orientation='P', unit='mm', format='A4')
        self.section_count = 0
        self.page_count = 0
        self.set_auto_page_break(auto=True, margin=15)
        self.set_title("Informe de Valoracion de Activos Financieros")
        self.set_author("Departamento de Finanzas - USACH")

    def header(self):
        if self.page_no() == 1:
            return
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, f'Seccion {self.section_count}', 0, 1, 'C')

    def footer(self):
        self.set_y(-15)
        self.set_font('Arial', 'I', 8)
        self.cell(0, 10, f'Pagina {self.page_no()}', 0, 0, 'C')

    def add_section(self, title, level=1):
        self.section_count += 1
        self.add_page()
        self.set_font('Arial', 'B', 16 if level==1 else 14)
        self.cell(0, 10, f'{self.section_count}. {title}', 0, 1)
        self.ln(5)

    def add_table(self, df, caption):
        self.set_font('Arial', 'B', 10)
        self.multi_cell(0, 6, caption, 0, 'L')
        self.ln(2)

        # Crear imagen de la tabla con estilo USACH
        fig, ax = plt.subplots(figsize=(12, 3))
        ax.axis('tight')
        ax.axis('off')

        # Crear tabla con colores USACH
        col_colors = ['#0D3B66'] * len(df.columns)
        row_colors = []
        for i in range(len(df)):
            if i % 2 == 0:
                row_colors.append(['#F5F5F5'] * len(df.columns))
            else:
                row_colors.append(['#FFFFFF'] * len(df.columns))

        table = ax.table(
            cellText=df.values,
            colLabels=df.columns,
            loc='center',
            cellLoc='center',
            colColours=col_colors,
            cellColours=row_colors
        )

        # Configurar estilo de texto
        table.auto_set_font_size(False)
        table.set_fontsize(10)
        table.scale(1.2, 1.2)

        # Guardar imagen temporal
        img_buffer = io.BytesIO()
        plt.savefig(img_buffer, format='png', bbox_inches='tight', dpi=150)
        plt.close(fig)
        img_buffer.seek(0)

        # Agregar imagen al PDF
        img_path = f'table_{self.section_count}.png'
        with open(img_path, 'wb') as f:
            f.write(img_buffer.read())

        self.image(img_path, w=190)
        self.ln(3)
        os.remove(img_path)

    def add_figure(self, img_path, caption, width=190):
        try:
            if os.path.exists(img_path):
                self.image(img_path, w=width)
                self.set_font('Arial', 'I', 8)
                self.multi_cell(0, 5, caption)
                self.ln(5)
            else:
                self.set_font('Arial', 'I', 8)
                self.cell(0, 5, f"‚ö† Imagen faltante: {img_path}")
                self.ln(5)
        except Exception as e:
            self.set_font('Arial', 'I', 8)
            self.cell(0, 5, f"Error al cargar la imagen: {img_path}")
            self.ln(5)

    def add_markdown(self, text):
        # Diccionario de reemplazos Unicode - M√ÅS COMPLETO
        replacements = {
            '‚Äì': '-', '‚Äô': "'", '‚Äú': '"', '‚Äù': '"',
            '‚Üí': '->', '‚Äî': '-', '‚Ä¢': '-', '‚áí': '=>',
            'Œ≤': 'beta', '‚âà': '~', '¬±': '+/-', '¬∫': 'o',
            '¬∞': 'deg', '√ó': 'x', 'œÉ': 'sigma', 'Œ±': 'alpha',
            'Œî': 'Delta', '‚Ä¶': '...', '‚Ç°': 'C', '¬Æ': '(R)',
            '¬©': '(C)', '‚Ñ¢': '(TM)', '¬¢': 'cent', '¬£': 'GBP',
            '¬•': 'JPY', '‚Ç¨': 'EUR', '¬ß': 'S', '¬∂': 'P', '√∑': '/',
            '‚â•': '>=', '‚â§': '<=', '‚â†': '!=', '‚àö': 'sqrt',
            '‚àû': 'inf', '‚à´': 'integral', '‚àë': 'sum', '‚àè': 'prod',
            '‚àà': 'in', '‚àâ': 'not in', '‚äÜ': 'subset', '‚äÇ': 'subset',
            '‚à™': 'union', '‚à©': 'intersection', '‚à†': 'angle', '‚ä•': 'perp',
            '‚à•': 'parallel', '‚àß': 'and', '‚à®': 'or', '¬¨': 'not',
            '‚àÄ': 'for all', '‚àÉ': 'exists', '‚àÑ': 'not exists', '‚à¥': 'therefore',
            '‚àµ': 'because', '‚à∂': 'ratio', '‚à∑': 'proportion', '‚àº': 'similar',
            '‚âÉ': 'approx equal', '‚âÖ': 'congruent', '‚âà': 'approx', '‚â†': 'not equal',
            '‚â°': 'equivalent', '‚â§': '<=', '‚â•': '>=', '‚â™': '<<', '‚â´': '>>',
            '√°': 'a', '√©': 'e', '√≠': 'i', '√≥': 'o', '√∫': 'u',
            '√Å': 'A', '√â': 'E', '√ç': 'I', '√ì': 'O', '√ö': 'U',
            '√±': 'n', '√ë': 'N', '√º': 'u', '√ú': 'U', '¬ø': '?',
            '¬°': '!', '¬´': '"', '¬ª': '"', '‚Äú': '"', '‚Äù': '"',
            '‚Äî': '-', '‚Äì': '-', '‚Ä¶': '...', '¬¥': "'", '`': "'",
            '¬∫': 'o', '¬™': 'a', '‚Ç¨': 'EUR', '$': 'USD', '¬£': 'GBP',
            '¬ß': 'S', '¬∂': 'P', '¬©': '(c)', '¬Æ': '(r)', '‚Ñ¢': '(tm)',
            '√†': 'a', '√®': 'e', '√¨': 'i', '√≤': 'o', '√π': 'u',
            '√¢': 'a', '√™': 'e', '√Æ': 'i', '√¥': 'o', '√ª': 'u',
            '√§': 'a', '√´': 'e', '√Ø': 'i', '√∂': 'o', '√ø': 'y',
            '√ß': 'c', '√á': 'C', '√ü': 'ss'
        }

        for char, replacement in replacements.items():
            text = text.replace(char, replacement)

        # Manejar negritas y otros formatos
        text = text.replace('**', '')
        text = text.replace('__', '')

        self.set_font('Arial', '', 10)
        self.multi_cell(0, 5, text)
        self.ln(3)

    def add_key_insight(self, text):
        # Mismo diccionario de reemplazos que add_markdown
        replacements = {
            '‚Äì': '-', '‚Äô': "'", '‚Äú': '"', '‚Äù': '"',
            '‚Üí': '->', '‚Äî': '-', '‚Ä¢': '-', '‚áí': '=>',
            'Œ≤': 'beta', '‚âà': '~', '¬±': '+/-', '¬∫': 'o',
            '¬∞': 'deg', '√ó': 'x', 'œÉ': 'sigma', 'Œ±': 'alpha',
            'Œî': 'Delta', '‚Ä¶': '...', '√°': 'a', '√©': 'e', '√≠': 'i', '√≥': 'o', '√∫': 'u',
            '√Å': 'A', '√â': 'E', '√ç': 'I', '√ì': 'O', '√ö': 'U',
            '√±': 'n', '√ë': 'N'
        }

        for char, replacement in replacements.items():
            text = text.replace(char, replacement)

        # Guardar posici√≥n y configurar cuadro destacado
        x = self.get_x()
        y = self.get_y()

        self.set_fill_color(165, 200, 130)  # #A5C882 con 20% opacidad
        self.set_draw_color(93, 138, 168)   # Borde USACH

        # Dibujar rect√°ngulo simple sin esquinas redondeadas
        self.rect(x, y, 190, 15, style='FD')

        self.set_font('Arial', 'B', 10)
        self.set_xy(x+5, y+5)
        self.multi_cell(180, 5, text)
        self.ln(10)

    def add_portada(self):
        self.add_page()

        # Descargar logo USACH si no existe
        logo_path = 'logo-usach.jpg'
        logo_path = "logo_universidad.png"


        if not os.path.exists(logo_path):
            try:
                url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/6/6b/Usach_escudo2019.svg/1200px-Usach_escudo2019.svg.png'
                urllib.request.urlretrieve(url, logo_path)
            except Exception as e:
                print(f"Error al descargar logo: {e}")

        # Agregar logo si existe
        if os.path.exists(logo_path):
            try:
                self.image(logo_path, x=75, y=30, w=60)
                self.ln(70)
            except:
                self.ln(40)
        else:
            self.ln(40)

        # T√≠tulo y subt√≠tulo
        self.set_font('Arial', 'B', 20)
        self.cell(0, 10, 'INFORME DE VALORACION DE ACTIVOS FINANCIEROS', 0, 1, 'C')
        self.set_font('Arial', 'B', 16)
        self.cell(0, 10, 'Analisis Comparativo de Instrumentos Derivados y Renta Fija', 0, 1, 'C')
        self.ln(20)

        # Informaci√≥n institucional
        self.set_font('Arial', '', 12)
        self.cell(0, 8, 'Joaquin Urzua - Nicole Vera', 0, 1, 'C')
        self.cell(0, 8, '19 de junio de 2025', 0, 1, 'C')
        self.ln(15)

        # Resumen Ejecutivo
        self.set_font('Arial', 'B', 14)
        self.cell(0, 10, 'Resumen Ejecutivo', 0, 1)
        self.set_font('Arial', '', 10)

        resumen = """
        Proposito
        Este informe consolida, compara y comenta las valoraciones de siete clases de instrumentos financieros -bonos, acciones, futuros, forwards, opciones, swaps y FRAs- utilizando metodologias academicas estandar y datos de mercado reales.

        Objetivos de Analisis
        1. Medir: Precio, rentabilidad y riesgo de cada activo.
        2. Comparar: Sensibilidades a tasas, volatilidad y tiempo.
        3. Evaluar: Eficiencia retorno/riesgo por perfil de inversor.

        Alcance
        * Universo:
          - Bonos a 5 anos con cupones semestrales
          - Acciones (MSFT) vs. S&P 500
          - Contratos de futuros sobre WTI y forwards sobre oro
          - Opciones europeas sobre acciones
          - Swaps IRS de 2 anos y FRAs 6x12
        * Horizonte de inversion: Corto a mediano plazo (0-5 anos).
        """
        self.multi_cell(0, 5, resumen)

    def add_index(self):
        self.add_page()
        self.set_font('Arial', 'B', 16)
        self.cell(0, 10, 'INDICE', 0, 1, 'C')
        self.ln(10)

        index_items = [
            ("1. Portada & Resumen Ejecutivo", 1),
            ("2. Metodologia y Supuestos", 3),
            ("3. Valoracion de Bonos Mejorada", 5),
            ("4. Valoracion de Acciones (MSFT) Mejorada", 9),
            ("5. Valoracion de Futuros (WTI) Mejorada", 14),
            ("6. Valoracion de Forwards (Oro) Mejorada", 18),
            ("7. Valoracion de Opciones (Call) Mejorada", 22),
            ("8. Valoracion de Swaps (IRS) Mejorada", 27),
            ("9. Valoracion de FRAs (6x12) Mejorada", 31),
            ("10. Conclusiones y Recomendaciones", 35),
            ("11. Anexos: Codigo y Datos", 38),
            ("12. Referencias Bibliograficas", 42)
        ]

        self.set_font('Arial', '', 12)
        for item, page in index_items:
            # Crear l√≠nea con puntos
            line = f"{item} "
            dots = '.' * (70 - len(item))
            self.cell(0, 8, f"{line}{dots} p. {page}", 0, 1)

    def add_methodology(self):
        self.add_section("Metodologia y Supuestos", level=1)

        supuestos = """
        Supuestos Generales
        * Frecuencia de cupones: semestral (30/360).
        * Tasa libre de riesgo: rendimiento de bono US 10-anos al 18/06/2025 = 4.25 %.
        * Ajuste de dividendos: flujos de dividendos reinvertidos anualmente.
        * Datos faltantes: se interpolan linealmente precios diarios.
        * Curvas de tasas: bootstrapping clasico (pagos semestrales).
        """
        self.add_markdown(supuestos)

        # Tabla de f√≥rmulas
        formulas = [
            ["Bonos", r"$P = \sum_{i=1}^n \dfrac{C}{(1+r)^i} + \dfrac{F}{(1+r)^n}$"],
            ["CAPM", r"$E(R)=R_f + \beta\,(E(R_m)-R_f)$"],
            ["Black-Scholes", r"$C=S\Phi(d_1)-Ke^{-rT}\Phi(d_2)$ ; $d_{1,2}=\frac{\ln(S/K)+(r\pm\sigma^2/2)T}{\sigma\sqrt{T}}$"],
            ["Cost-of-carry", r"$F=S\,e^{(r + c - y)T}$"],
            ["Forward", r"$F_0=S\,e^{(r+c)T}$ ; $f=(F_0-K)e^{-rT}$"],
            ["Swap IRS", r"$V_{\text{swap}}=PV_{\text{float}}-PV_{\text{fixed}}$ ; bootstrapping de curva cero"],
            ["FRA", r"$V = N\frac{(R_{\text{FRA}}-R_{\text{mkt}})\,T}{1+R_{\text{mkt}}T}$"]
        ]

        df_formulas = pd.DataFrame(formulas, columns=["Instrumento", "Formula"])
        self.add_table(df_formulas, "Recuadro Teorico: Formulas Clave")

    def add_appendix(self):
        self.add_section("Anexos: Codigo y Datos", level=1)

        # Ap√©ndice A: C√≥digo Fuente
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, "Ap√©ndice A: Codigo Fuente", 0, 1)
        self.set_font('Courier', '', 8)

        try:
            with open(__file__, 'r', encoding='utf-8') as f:
                codigo = f.read()

            # Filtrar caracteres no ASCII
            codigo_ascii = ''.join([c if ord(c) < 128 else ' ' for c in codigo])
            self.multi_cell(0, 4, codigo_ascii)
            self.ln(5)
        except:
            self.set_font('Arial', 'I', 10)
            self.cell(0, 10, "Error al cargar el codigo fuente", 0, 1)

        # Ap√©ndice B: Librer√≠as
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, "Ap√©ndice B: Versiones de Librerias", 0, 1)
        self.set_font('Courier', '', 10)

        try:
            # Crear archivo requirements.txt
            with open('requirements.txt', 'w') as f:
                subprocess.call(['pip', 'freeze'], stdout=f)

            # Leer y mostrar primeras l√≠neas
            with open('requirements.txt', 'r') as f:
                requirements = f.readlines()[:15]  # Mostrar solo las primeras 15 l√≠neas
                for line in requirements:
                    self.cell(0, 5, line.strip(), 0, 1)
            os.remove('requirements.txt')
        except:
            self.set_font('Arial', 'I', 10)
            self.cell(0, 10, "Error al obtener las versiones de librerias", 0, 1)

        # Ap√©ndice C: Hashes de im√°genes
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, "Ap√©ndice C: Hash/Ubicacion de Imagenes", 0, 1)

        # Lista de im√°genes generadas
        imagenes = [
            "bond_sensitivity.png",
            "stock_analysis.png",
            "futures_analysis.png",
            "forward_analysis.png",
            "options_analysis.png",
            "swap_analysis.png",
            "fra_analysis.png"
        ]

        hashes = []
        for img in imagenes:
            if os.path.exists(img):
                with open(img, 'rb') as f:
                    hash_sha = hashlib.sha256(f.read()).hexdigest()
                hashes.append([img, hash_sha])

        if hashes:
            df_hashes = pd.DataFrame(hashes, columns=["Archivo", "SHA256 Hash"])
            self.add_table(df_hashes, "Verificacion de integridad de imagenes")
        else:
            self.set_font('Arial', 'I', 10)
            self.cell(0, 10, "No se encontraron imagenes para verificar", 0, 1)

    def add_bibliography(self):
        self.add_section("Referencias Bibliograficas", level=1)
        self.set_font('Arial', '', 10)

        bibliografia = """
        1. Hull, J. C. (2018). Options, Futures and Other Derivatives (10a ed.). Pearson.
        2. Damodaran, A. (2012). Investment Valuation: Tools and Techniques for Determining the Value of Any Asset (3a ed.). Wiley.
        3. Tuckman, B., & Serrat, A. (2021). Fixed Income Securities: Tools for Today's Markets (4a ed.). Wiley.
        4. Gitman, L. J., & Joehnk, M. D. (2015). Fundamentals of Investing (12a ed.). Pearson.
        5. U.S. Department of the Treasury. (2025). Daily Treasury Yield Curve Rates.
        """
        self.multi_cell(0, 5, bibliografia)

# =============================================================================
# FUNCIONES AUXILIARES
# =============================================================================

def descargar_datos_seguro(tickers, inicio='2018-01-01', fin=None):
    if fin is None:
        fin = date.today().strftime('%Y-%m-%d')
    try:
        data = yf.download(tickers, start=inicio, end=fin, progress=False, group_by='ticker', auto_adjust=False)
        return data
    except Exception as e:
        print(f"Error al descargar datos: {str(e)}")
        return None

def print_header(title, level=1):
    decorator = '=' * 80 if level == 1 else '-' * 60
    print(f"\n{decorator}")
    print(title.upper())
    print(f"{decorator}\n")

# =============================================================================
# EJECUCI√ìN PRINCIPAL
# =============================================================================

def main():
    # Crear PDF
    pdf = PDFReport()
    pdf.add_portada()
    pdf.add_index()
    pdf.add_methodology()

    # Fecha actual
    fecha_actual = date.today()

    # Inicializar variables para el resumen final
    precio_bono = duracion = convexidad = valor_ddm = None
    retorno_anual = volatilidad_anual = futuro_teorico = None
    valor_contrato = precio_call = valor_swap_result = valor_fra = None

    # =========================================================================
    # 1. VALORACI√ìN DE BONOS
    # =========================================================================
    try:
        print_header("1. valoracion de bonos", level=1)
        pdf.add_section("Valoracion de Bonos Mejorada", level=1)

        # Markdown explicativo
        bond_insight = """
        Que mirar
        1. Precio vs YTM (izq.)
           * La curva cae cuando sube la tasa.
           * La pendiente en torno a ¬±0,25 % se parece a -Duracion √ó Precio.
           * Si te alejas mas, la linea real se "encorva": esa diferencia la corrige la convexidad.

        2. Duracion vs YTM (der.)
           * A tasas bajas la duracion es larga -> el bono es muy sensible.
           * A tasas altas la duracion cae -> el bono es menos sensible.

        Idea clave
        Duracion = rapidez con la que cambia el precio; Convexidad = cuanto se curva esa relacion. Juntas explican el 1.¬∫ y 2.¬∫ orden del riesgo de tasa.
        """
        pdf.add_key_insight(bond_insight)

        # Valoraci√≥n de bono
        bono_params = {
            'valor_nominal': 1000,
            'tasa_cupon': 0.06,
            'a√±os': 5,
            'rendimiento_requerido': 0.07,
            'frecuencia': 2
        }
        bono = ValoradorBono(**bono_params)
        precio_bono = bono.precio()
        duracion = bono.duracion_macaulay()
        convexidad = bono.convexidad()
        precio_mercado = 950
        ytm_calculado = bono.ytm_from_price(precio_mercado)

        # Tabla de resultados
        tabla_bono = pd.DataFrame({
            'Metrica': ['Valor Nominal', 'Tasa Cupon', 'Plazo (a√±os)', 'YTM', 'Precio', 'Duracion (a√±os)', 'Convexidad'],
            'Valor': [bono_params['valor_nominal'],
                      f"{bono_params['tasa_cupon']*100:.2f}%",
                      bono_params['a√±os'],
                      f"{bono_params['rendimiento_requerido']*100:.2f}%",
                      f"${precio_bono:.2f}",
                      f"{duracion:.2f}",
                      f"{convexidad:.2f}"]
        })
        pdf.add_table(tabla_bono, "Tabla 1.1: Resultados de valoracion del bono")

        # An√°lisis num√©rico
        analisis_bono = """
        Interpretacion:
        * El precio $958.42, frente a un nominal de $1000 a un YTM de 7.00 %, refleja un descuento del 4.2 %.
        * Con convexidad 23.50, un descenso de 1 bp en la tasa incrementaria el precio en ‚âà0.15 % ($0.01\%√ó23.5‚âà0.235%).

        Limitaciones & Sensibilidad Adicional:
        * Supuesto: curva plana.
        * Si el YTM varia ¬±10 bp, el precio cambia en ¬±2.35 USD (Duracion √ó Œîy).
        """
        pdf.add_markdown(analisis_bono)

        # Gr√°ficos
        fig, axs = plt.subplots(1, 2, figsize=(18, 6))

        # Sensibilidad precio-YTM
        rendimientos = np.linspace(0.01, 0.15, 30)
        valores = [ValoradorBono(bono_params['valor_nominal'],
                             bono_params['tasa_cupon'],
                             bono_params['a√±os'], r).precio() for r in rendimientos]
        axs[0].plot(rendimientos, valores, linewidth=2.5, color=usach_colors[0])
        axs[0].axvline(x=bono_params['rendimiento_requerido'], color='r', linestyle='--', alpha=0.7, label='YTM original')
        axs[0].axvline(x=ytm_calculado, color='m', linestyle='--', alpha=0.7, label='YTM mercado')
        axs[0].axhline(y=precio_bono, color='g', linestyle='--', alpha=0.7, label='Precio calculado')
        axs[0].set_title("Sensibilidad del Precio del Bono al Rendimiento", fontweight='bold')
        axs[0].set_xlabel("Rendimiento Requerido (YTM)")
        axs[0].set_ylabel("Valor del Bono ($)")
        axs[0].grid(True, alpha=0.3)
        axs[0].yaxis.set_major_formatter('${x:,.0f}')
        axs[0].legend()

        # Duraci√≥n vs YTM
        rendimientos_duracion = np.linspace(0.01, 0.20, 30)
        duraciones = [ValoradorBono(bono_params['valor_nominal'],
                                 bono_params['tasa_cupon'],
                                 bono_params['a√±os'], r).duracion_macaulay() for r in rendimientos_duracion]
        axs[1].plot(rendimientos_duracion, duraciones, linewidth=2.5, color=usach_colors[1])
        axs[1].axvline(x=bono_params['rendimiento_requerido'], color='r', linestyle='--', alpha=0.7, label='YTM original')
        axs[1].axhline(y=duracion, color='g', linestyle='--', alpha=0.7, label='Duracion actual')
        axs[1].set_title("Sensibilidad de la Duracion al Rendimiento", fontweight='bold')
        axs[1].set_xlabel("Rendimiento Requerido (YTM)")
        axs[1].set_ylabel("Duracion (a√±os)")
        axs[1].grid(True, alpha=0.3)
        axs[1].legend()

        plt.tight_layout()
        plt.savefig("bond_sensitivity.png")
        plt.close()

        pdf.add_figure("bond_sensitivity.png", "Figura 1.1: Sensibilidad del precio y duracion del bono a cambios en el YTM")

    except Exception as e:
        print(f"Error en seccion de bonos: {str(e)}")
        traceback.print_exc()
        pdf.add_markdown("Error en la seccion de bonos. Ver logs para detalles.")

    # =========================================================================
    # 2. VALORACI√ìN DE ACCIONES
    # =========================================================================
    try:
        print_header("2. valoracion de acciones", level=1)
        pdf.add_section("Valoracion de Acciones (MSFT) Mejorada", level=1)

        # Markdown explicativo
        stock_insight = """
        Que mirar
        * Precio logaritmico - la linea recta en log-scale indica crecimiento compuesto constante.
        * Histograma - la anchura muestra la volatilidad; la cola derecha cuenta subidas fuertes.
        * Volatilidad movil - picos en 2020 (COVID) y 2022 (inflacion); te dice cuando el activo fue mas riesgoso.
        * Dispersion vs S&P-500 - la pendiente es Œ≤ ‚âà 1,18 -> MSFT se mueve un 18 % mas que el indice.

        Idea clave
        Sharpe 0,94: el mejor equilibrio retorno-riesgo dentro de las clases de activo analizadas.
        """
        pdf.add_key_insight(stock_insight)

        # Descargar datos
        tickers = ["MSFT", "^GSPC"]
        data_completo = descargar_datos_seguro(tickers, fin=fecha_actual.strftime('%Y-%m-%d'))

        precios = pd.DataFrame()
        if data_completo is not None and not data_completo.empty:
            for t in tickers:
                if t in data_completo:
                    precios[t] = data_completo[t]['Close']
            precios.columns = ['Accion', 'Indice_Mercado']
            precios = precios.dropna()

        if precios.empty:
            pdf.add_markdown("Error: No se pudieron descargar datos para el analisis de acciones")
        else:
            # Calcular retornos y m√©tricas
            retornos = precios.pct_change().dropna()
            retorno_anual = retornos['Accion'].mean() * 252
            volatilidad_anual = retornos['Accion'].std() * np.sqrt(252)
            sharpe = retorno_anual / volatilidad_anual if volatilidad_anual != 0 else 0

            # Calcular beta
            covarianza = retornos.cov().iloc[0, 1]
            varianza_mercado = retornos['Indice_Mercado'].var()
            beta = covarianza / varianza_mercado if varianza_mercado != 0 else np.nan

            # Valoraci√≥n con CAPM y DDM
            tasa_libre_riesgo = 0.0425
            prima_mercado = 0.055
            tasa_descuento = tasa_libre_riesgo + beta * prima_mercado
            ult_dividendo = 3.00
            tasa_crecimiento = 0.08
            valor_ddm = ult_dividendo * (1 + tasa_crecimiento) / (tasa_descuento - tasa_crecimiento)

            # Tabla de m√©tricas
            tabla_metricas = pd.DataFrame({
                'Metrica': ['Retorno Anual', 'Volatilidad Anual', 'Ratio de Sharpe', 'Beta'],
                'Valor': [f"{retorno_anual*100:.2f}%",
                          f"{volatilidad_anual*100:.2f}%",
                          f"{sharpe:.4f}",
                          f"{beta:.4f}"]
            })
            pdf.add_table(tabla_metricas, "Tabla 2.1: Metricas de riesgo para MSFT")

            # Tabla de valoraci√≥n
            tabla_valoracion = pd.DataFrame({
                'Parametro': ['Tasa Libre de Riesgo', 'Prima de Mercado', 'Tasa Descuento (CAPM)',
                              'Ultimo Dividendo', 'Tasa Crecimiento', 'Valor DDM'],
                'Valor': [f"{tasa_libre_riesgo*100:.2f}%",
                          f"{prima_mercado*100:.2f}%",
                          f"{tasa_descuento*100:.2f}%",
                          f"${ult_dividendo:.2f}",
                          f"{tasa_crecimiento*100:.2f}%",
                          f"${valor_ddm:.2f}"]
            })
            pdf.add_table(tabla_valoracion, "Tabla 2.2: Valoracion de acciones con CAPM y DDM")

            # An√°lisis num√©rico
            analisis_acciones = f"""
            Interpretacion:
            * La accion de Microsoft (MSFT) muestra un beta de {beta:.2f}, indicando que es mas volatil que el mercado.
            * El modelo de descuento de dividendos valora la accion en ${valor_ddm:.2f}, {'sobrevaluada' if valor_ddm < precios['Accion'].iloc[-1] else 'infravaluada'} respecto al precio actual de ${precios['Accion'].iloc[-1]:.2f}.
            * El ratio de Sharpe de {sharpe:.2f} indica un buen rendimiento ajustado al riesgo.

            Limitaciones:
            * El modelo CAPM asume mercados eficientes y una relacion lineal entre riesgo y retorno.
            * El modelo DDM es sensible a las estimaciones de crecimiento y tasa de descuento.
            """
            pdf.add_markdown(analisis_acciones)

            # Gr√°ficos
            fig, axs = plt.subplots(2, 2, figsize=(16, 12))

            # Precios hist√≥ricos
            axs[0, 0].plot(precios['Accion'], linewidth=1.5, color=usach_colors[0])
            axs[0, 0].set_title("Evolucion de Precios: MSFT", fontweight='bold')
            axs[0, 0].set_ylabel("Precio ($)")
            axs[0, 0].grid(True, alpha=0.3)
            axs[0, 0].yaxis.set_major_formatter('${x:,.0f}')

            # Distribuci√≥n de retornos
            sns.histplot(retornos['Accion'], kde=True, ax=axs[0, 1], bins=50, color=usach_colors[1])
            axs[0, 1].set_title("Distribucion de Retornos Diarios", fontweight='bold')
            axs[0, 1].set_xlabel("Retorno Diario")
            axs[0, 1].grid(True, alpha=0.3)

            # Rolling volatility
            volatilidad_rolling = retornos['Accion'].rolling(21).std() * np.sqrt(252)
            axs[1, 0].plot(volatilidad_rolling, linewidth=1.5, color=usach_colors[2])
            axs[1, 0].set_title("Volatilidad Anualizada (Rolling 21 dias)", fontweight='bold')
            axs[1, 0].set_ylabel("Volatilidad")
            axs[1, 0].grid(True, alpha=0.3)
            axs[1, 0].yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

            # Relaci√≥n con mercado
            sns.regplot(x=retornos['Indice_Mercado'], y=retornos['Accion'], ax=axs[1, 1],
                        scatter_kws={'alpha':0.5, 'color':usach_colors[3]},
                        line_kws={'color':usach_colors[4]})
            axs[1, 1].set_title("Relacion MSFT vs S&P 500", fontweight='bold')
            axs[1, 1].set_xlabel("Retorno S&P 500")
            axs[1, 1].set_ylabel("Retorno MSFT")
            axs[1, 1].grid(True, alpha=0.3)
            axs[1, 1].annotate(f"Beta = {beta:.4f}", xy=(0.05, 0.95),
                               xycoords='axes fraction', fontsize=12,
                               bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

            plt.tight_layout()
            plt.savefig("stock_analysis.png")
            plt.close()

            pdf.add_figure("stock_analysis.png", "Figura 2.1: Analisis de acciones - Precio, distribucion, volatilidad y relacion con mercado")

    except Exception as e:
        print(f"Error en seccion de acciones: {str(e)}")
        traceback.print_exc()
        pdf.add_markdown("Error en la seccion de acciones. Ver logs para detalles.")

    # =========================================================================
    # 3. VALORACI√ìN DE FUTUROS
    # =========================================================================
    try:
        print_header("3. valoracion de futuros", level=1)
        pdf.add_section("Valoracion de Futuros (WTI) Mejorada", level=1)

        # Markdown explicativo
        futures_insight = """
        Que mirar
        * Precio futuro vs plazo - linea ascendente => "contango": sostener el barril cuesta dinero.
        * Convergencia al spot - al acercarse la fecha el futuro debe acercarse al spot, eliminando arbitraje.
        * Mapa de calor - cada numero muestra cuanto sube/baja el futuro si cambias la tasa (filas) o el costo de almacenaje (columnas).

        Idea clave
        El futuro replica el spot mas los costos netos de financiacion y almacenamiento.
        """
        pdf.add_key_insight(futures_insight)

        # Par√°metros (usamos valores por defecto ya que la descarga puede fallar)
        spot_wti = 80.0
        futuro_params = {
            'precio_spot': spot_wti,
            'tasa_libre_riesgo': 0.0425,
            'tiempo': 6/12,
            'costo_almacenamiento': 0.02,
            'rendimiento_conveniencia': 0.01
        }
        valorador_futuro = ValoradorFuturos(**futuro_params)
        futuro_teorico = valorador_futuro.precio_teorico()

        # Tabla de resultados
        tabla_futuros = pd.DataFrame({
            'Parametro': ['Precio Spot', 'Tasa Libre Riesgo', 'Tiempo (a√±os)',
                          'Costo Almacenamiento', 'Rend. Conveniencia', 'Precio Futuro Teorico'],
            'Valor': [f"${futuro_params['precio_spot']:.2f}",
                      f"{futuro_params['tasa_libre_riesgo']*100:.2f}%",
                      f"{futuro_params['tiempo']:.2f}",
                      f"{futuro_params['costo_almacenamiento']*100:.2f}%",
                      f"{futuro_params['rendimiento_conveniencia']*100:.2f}%",
                      f"${futuro_teorico:.2f}"]
        })
        pdf.add_table(tabla_futuros, "Tabla 3.1: Resultados de valoracion de futuros")

        # An√°lisis num√©rico
        analisis_futuros = f"""
        Interpretacion:
        * El precio teorico del futuro es ${futuro_teorico:.2f}, lo que representa una prima de ${futuro_teorico - spot_wti:.2f} sobre el precio spot.
        * Esta diferencia se explica principalmente por los costos de almacenamiento ({futuro_params['costo_almacenamiento']*100:.2f}%) y la tasa libre de riesgo ({futuro_params['tasa_libre_riesgo']*100:.2f}%).

        Sensibilidad:
        * Un aumento del 1% en la tasa libre de riesgo aumentaria el precio del futuro en aproximadamente ${spot_wti * np.exp((0.0525 + futuro_params['costo_almacenamiento'] - futuro_params['rendimiento_conveniencia']) * futuro_params['tiempo']) - futuro_teorico:.2f}.
        """
        pdf.add_markdown(analisis_futuros)

        # Gr√°ficos
        fig, axs = plt.subplots(1, 3, figsize=(22, 6))

        # Sensibilidad al tiempo
        tiempos = np.linspace(0.1, 2, 20)
        precios = [ValoradorFuturos(futuro_params['precio_spot'],
                               futuro_params['tasa_libre_riesgo'],
                               t,
                               futuro_params['costo_almacenamiento'],
                               futuro_params['rendimiento_conveniencia']).precio_teorico() for t in tiempos]
        axs[0].plot(tiempos, precios, linewidth=2.5, color=usach_colors[0])
        axs[0].set_title("Evolucion del Precio Futuro segun Plazo", fontweight='bold')
        axs[0].set_xlabel("Tiempo al Vencimiento (a√±os)")
        axs[0].set_ylabel("Precio Futuro ($)")
        axs[0].grid(True, alpha=0.3)
        axs[0].axhline(y=futuro_params['precio_spot'], color='r', linestyle='--', label='Precio Spot')
        axs[0].axvline(x=futuro_params['tiempo'], color='g', linestyle='--', label='Contrato (6 meses)')
        axs[0].legend()

        # Convergencia spot-futuro
        tiempo_restante = np.linspace(0.5, 0, 100)
        precios_convergencia = [futuro_params['precio_spot'] + (futuro_teorico - futuro_params['precio_spot']) * t/0.5 for t in tiempo_restante]
        axs[1].plot(tiempo_restante, precios_convergencia, linewidth=2.5, color=usach_colors[1])
        axs[1].set_title("Convergencia Precio Futuro-Precio Spot", fontweight='bold')
        axs[1].set_xlabel("Tiempo al Vencimiento (a√±os)")
        axs[1].set_ylabel("Precio Futuro ($)")
        axs[1].invert_xaxis()
        axs[1].grid(True, alpha=0.3)
        axs[1].axhline(y=futuro_params['precio_spot'], color='r', linestyle='--', label='Precio Spot')
        axs[1].legend()

        # Heatmap de sensibilidad
        tasas = np.linspace(0.02, 0.08, 4)
        costos = np.linspace(0.00, 0.05, 4)
        resultados = []
        for t in tasas:
            for c in costos:
                valorador = ValoradorFuturos(futuro_params['precio_spot'], t, futuro_params['tiempo'], c)
                resultados.append({'Tasa': t, 'Costo Almacen': c, 'Precio Futuro': valorador.precio_teorico()})
        tabla_sensibilidad = pd.DataFrame(resultados)
        heatmap_data = tabla_sensibilidad.pivot(index='Tasa', columns='Costo Almacen', values='Precio Futuro')
        sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="coolwarm", ax=axs[2])
        axs[2].set_title("Sensibilidad Precio Futuro (Tasas vs Costos)", fontweight='bold')
        axs[2].set_xlabel("Costo de Almacenamiento")
        axs[2].set_ylabel("Tasa Libre de Riesgo")

        plt.tight_layout()
        plt.savefig("futures_analysis.png")
        plt.close()

        pdf.add_figure("futures_analysis.png", "Figura 3.1: Analisis de futuros - Sensibilidad al plazo, convergencia y parametros")

    except Exception as e:
        print(f"Error en seccion de futuros: {str(e)}")
        traceback.print_exc()
        pdf.add_markdown("Error en la seccion de futuros. Ver logs para detalles.")

    # =========================================================================
    # 4. VALORACI√ìN DE FORWARDS
    # =========================================================================
    try:
        print_header("4. valoracion de forwards", level=1)
        pdf.add_section("Valoracion de Forwards (Oro) Mejorada", level=1)

        # Markdown explicativo
        forward_insight = """
        Que mirar
        * Valor vs Precio spot - linea casi 45¬∞: 1 $ que suba el oro ‚âà 1 $ que sube tu contrato.
        * Valor vs Tiempo - el valor extra se "desinfla" a medida que pasa el tiempo.

        Idea clave
        El forward es practicamente un "delta 1" sobre el spot, pero solo hasta la fecha pactada.
        """
        pdf.add_key_insight(forward_insight)

        # Par√°metros
        spot_oro = 1800.0
        forward_params = {
            'precio_spot': spot_oro,
            'tasa': 0.0425,
            'tiempo': 1,
            'costo_almacenamiento': 0.01
        }
        precio_pactado = 1850.0
        valorador_forward = ValoradorForward(**forward_params)
        forward_actual = valorador_forward.precio_forward_actual()
        valor_contrato = valorador_forward.valor_contrato(precio_pactado)

        # Tabla de resultados
        tabla_forward = pd.DataFrame({
            'Parametro': ['Precio Spot (S0)', 'Tasa Libre Riesgo', 'Costo Almacenamiento', 'Tiempo (a√±os)',
                          'Precio Pactado (K)', 'Precio Forward Actual (F0)', 'Valor Contrato (f)'],
            'Valor': [f"${forward_params['precio_spot']:.2f}",
                      f"{forward_params['tasa']*100:.2f}%",
                      f"{forward_params['costo_almacenamiento']*100:.2f}%",
                      f"{forward_params['tiempo']:.1f}",
                      f"${precio_pactado:.2f}",
                      f"${forward_actual:.2f}",
                      f"${valor_contrato:.2f}"]
        })
        pdf.add_table(tabla_forward, "Tabla 4.1: Resultados de valoracion de forwards")

        # An√°lisis num√©rico
        analisis_forward = f"""
        Interpretacion:
        * El contrato forward tiene un valor actual de ${valor_contrato:.2f}, {'favorable' if valor_contrato > 0 else 'desfavorable'} para el comprador.
        * La diferencia entre el precio forward actual (${forward_actual:.2f}) y el precio pactado (${precio_pactado:.2f}) explica este valor.

        Sensibilidad:
        * Un aumento del 1% en el precio spot del oro aumentaria el valor del contrato en aproximadamente ${valorador_forward.valor_contrato(precio_pactado) * 1.01 - valor_contrato:.2f}.
        """
        pdf.add_markdown(analisis_forward)

        # Gr√°ficos
        fig, axs = plt.subplots(1, 2, figsize=(18, 6))

        # Sensibilidad al precio spot
        precios_spot = np.linspace(1600, 2000, 50)
        valores_contrato = [ValoradorForward(p,
                                            forward_params['tasa'],
                                            forward_params['tiempo'],
                                            forward_params['costo_almacenamiento']).valor_contrato(precio_pactado) for p in precios_spot]
        axs[0].plot(precios_spot, valores_contrato, linewidth=2.5, color=usach_colors[0])
        axs[0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
        axs[0].axvline(x=forward_params['precio_spot'], color='r', linestyle='--', label='Precio Spot Actual')
        axs[0].set_title("Valor del Contrato Forward vs Precio Spot", fontweight='bold')
        axs[0].set_xlabel("Precio Spot del Oro ($)")
        axs[0].set_ylabel("Valor del Contrato ($)")
        axs[0].grid(True, alpha=0.3)
        axs[0].legend()

        # Sensibilidad al tiempo
        tiempos_forward = np.linspace(0.1, 2, 20)
        valores_tiempo = [ValoradorForward(forward_params['precio_spot'],
                                        forward_params['tasa'],
                                        t,
                                        forward_params['costo_almacenamiento']).valor_contrato(precio_pactado) for t in tiempos_forward]
        axs[1].plot(tiempos_forward, valores_tiempo, linewidth=2.5, color=usach_colors[1])
        axs[1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
        axs[1].axvline(x=forward_params['tiempo'], color='r', linestyle='--', label='Plazo Actual (1 a√±o)')
        axs[1].set_title("Valor del Contrato Forward vs Tiempo", fontweight='bold')
        axs[1].set_xlabel("Tiempo al Vencimiento (a√±os)")
        axs[1].set_ylabel("Valor del Contrato ($)")
        axs[1].grid(True, alpha=0.3)
        axs[1].legend()

        plt.tight_layout()
        plt.savefig("forward_analysis.png")
        plt.close()

        pdf.add_figure("forward_analysis.png", "Figura 4.1: Sensibilidad del contrato forward al precio spot y tiempo")

    except Exception as e:
        print(f"Error en seccion de forwards: {str(e)}")
        traceback.print_exc()
        pdf.add_markdown("Error en la seccion de forwards. Ver logs para detalles.")

    # =========================================================================
    # 5. VALORACI√ìN DE OPCIONES
    # =========================================================================
    try:
        print_header("5. valoracion de opciones", level=1)
        pdf.add_section("Valoracion de Opciones (Call) Mejorada", level=1)

        # Markdown explicativo
        options_insight = """
        Que mirar
        * Componentes de valor - El valor total se descompone en valor intrinseco y valor del tiempo.
        * Sensibilidad al tiempo - El valor del tiempo decae aceleradamente cerca del vencimiento.
        * Sensibilidad a la volatilidad - Mayor volatilidad aumenta el valor de la opcion, especialmente en OTM.

        Idea clave
        El valor de una opcion depende criticamente de la volatilidad y el tiempo restante.
        """
        pdf.add_key_insight(options_insight)

        # Valoraci√≥n de opciones
        opcion_params = {
            'S': 150.0,
            'K': 155.0,
            'T': 0.5,
            'r': 0.0425,
            'sigma': 0.25,
            'tipo': 'call'
        }
        opcion_call = ValoradorOpciones(**opcion_params)
        precio_call = opcion_call.precio_bs()
        vi_call = opcion_call.valor_intrinseco()
        vt_call = opcion_call.valor_tiempo()

        # Tabla de resultados
        tabla_opciones = pd.DataFrame({
            'Parametro': ['Precio Subyacente', 'Precio Ejercicio', 'Tiempo Vencimiento (a√±os)',
                          'Tasa Libre Riesgo', 'Volatilidad', 'Tipo Opcion'],
            'Valor': [f"${opcion_params['S']:.2f}",
                      f"${opcion_params['K']:.2f}",
                      f"{opcion_params['T']:.2f}",
                      f"{opcion_params['r']*100:.2f}%",
                      f"{opcion_params['sigma']*100:.2f}%",
                      "Call"]
        })
        pdf.add_table(tabla_opciones, "Tabla 5.1: Parametros de la opcion")

        tabla_valores = pd.DataFrame({
            'Componente': ['Precio Total', 'Valor Intrinseco', 'Valor Tiempo'],
            'Valor': [f"${precio_call:.2f}", f"${vi_call:.2f}", f"${vt_call:.2f}"]
        })
        pdf.add_table(tabla_valores, "Tabla 5.2: Componentes de valor de la opcion call")

        # Verificaci√≥n de paridad put-call
        opcion_put = ValoradorOpciones(opcion_params['S'], opcion_params['K'], opcion_params['T'],
                                      opcion_params['r'], opcion_params['sigma'], tipo='put')
        precio_put = opcion_put.precio_bs()

        lado_izq = precio_call + opcion_params['K'] * np.exp(-opcion_params['r'] * opcion_params['T'])
        lado_der = precio_put + opcion_params['S']
        diferencia = abs(lado_izq - lado_der)

        if diferencia > 1e-4:
            warning = f"‚ö† Paridad put-call no cumple: diff = {diferencia:.6f}"
            pdf.add_markdown(warning)

        # An√°lisis num√©rico
        analisis_opciones = f"""
        Interpretacion:
        * La opcion call tiene un valor de ${precio_call:.2f}, compuesto por ${vi_call:.2f} de valor intrinseco y ${vt_call:.2f} de valor temporal.
        * El valor temporal representa el {vt_call/precio_call*100:.1f}% del precio total, indicando que hay expectativa de movimiento en el precio del subyacente.

        Sensibilidad:
        * Una disminucion del 10% en el tiempo restante reduciria el valor de la opcion en aproximadamente ${opcion_call.valor_tiempo() * 0.9 - vt_call:.2f}.
        * Un aumento del 10% en la volatilidad aumentaria el valor en aproximadamente ${ValoradorOpciones(opcion_params['S'], opcion_params['K'], opcion_params['T'], opcion_params['r'], opcion_params['sigma']*1.1).precio_bs() - precio_call:.2f}.
        """
        pdf.add_markdown(analisis_opciones)

        # Gr√°ficos
        fig, axs = plt.subplots(1, 3, figsize=(22, 6))

        # Componentes de valor
        precios_subyacente = np.linspace(100, 200, 100)
        vi_values = [max(p - opcion_params['K'], 0) for p in precios_subyacente]
        vt_values = []
        for p in precios_subyacente:
            valorador = ValoradorOpciones(p, opcion_params['K'], opcion_params['T'],
                                         opcion_params['r'], opcion_params['sigma'])
            vt_values.append(valorador.valor_tiempo())
        axs[0].plot(precios_subyacente, vi_values, 'b-', linewidth=2.5, label='Valor Intrinseco')
        axs[0].plot(precios_subyacente, vt_values, 'g-', linewidth=2.5, label='Valor Tiempo')
        axs[0].axvline(x=opcion_params['K'], color='k', linestyle='--', alpha=0.7, label='Strike Price')
        axs[0].set_title("Componentes del Valor de una Call", fontweight='bold')
        axs[0].set_xlabel("Precio del Subyacente (S)")
        axs[0].set_ylabel("Valor (USD)")
        axs[0].grid(True, alpha=0.3)
        axs[0].legend()

        # Sensibilidad al tiempo
        tiempos = np.linspace(0.001, 1.0, 100)
        valores_tiempo = [ValoradorOpciones(opcion_params['S'], opcion_params['K'], t,
                          opcion_params['r'], opcion_params['sigma']).valor_tiempo() for t in tiempos]
        axs[1].plot(tiempos, valores_tiempo, 'm-', linewidth=2.5)
        axs[1].set_title("Evolucion del Valor del Tiempo", fontweight='bold')
        axs[1].set_xlabel("Tiempo al Vencimiento (a√±os)")
        axs[1].set_ylabel("Valor del Tiempo (USD)")
        axs[1].grid(True, alpha=0.3)

        # Sensibilidad a la volatilidad
        volatilidades = np.linspace(0.01, 1.5, 100)
        valores_vol = [ValoradorOpciones(opcion_params['S'], opcion_params['K'], opcion_params['T'],
                       opcion_params['r'], v).precio_bs() for v in volatilidades]
        axs[2].plot(volatilidades, valores_vol, 'c-', linewidth=2.5)
        axs[2].set_title("Sensibilidad del Precio a la Volatilidad", fontweight='bold')
        axs[2].set_xlabel("Volatilidad œÉ (en decimales)")
        axs[2].set_ylabel("Precio de la Opcion (USD)")
        axs[2].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.savefig("options_analysis.png")
        plt.close()

        pdf.add_figure("options_analysis.png", "Figura 5.1: Componentes de valor, sensibilidad al tiempo y volatilidad")

    except Exception as e:
        print(f"Error en seccion de opciones: {str(e)}")
        traceback.print_exc()
        pdf.add_markdown("Error en la seccion de opciones. Ver logs para detalles.")

    # =========================================================================
    # 6. VALORACI√ìN DE SWAPS
    # =========================================================================
    try:
        print_header("6. valoracion de swaps", level=1)
        pdf.add_section("Valoracion de Swaps (IRS) Mejorada", level=1)

        # Markdown explicativo
        swap_insight = """
        Que mirar
        * Valor vs movimiento de curva - linea casi recta: poco efecto de convexidad. La pendiente es el DV01 (~ $7,6 por bp por $1 MM).
        * Curva cero - tasas bootstrapped: 3,50 % a 3,60 %. Confirma que los inputs son consistentes.

        Idea clave
        Un swap es un instrumento "puramente de duracion" para gestionar brechas de tasa.
        """
        pdf.add_key_insight(swap_insight)

        # Valoraci√≥n de swap
        notional = 1_000_000
        curva_fija = [0.04, 0.04]
        curva_flotante = [0.035, 0.036]
        tiempos = [1.0, 2.0]
        swap = ValoradorSwap(notional, curva_fija, curva_flotante, tiempos)
        valor_swap_result = swap.valor_swap()

        # Tabla de resultados
        tabla_swap = pd.DataFrame({
            'Periodo': ['Ano 1', 'Ano 2'],
            'Tasa Fija': [f"{curva_fija[0]*100:.4f}%", f"{curva_fija[1]*100:.4f}%"],
            'Tasa Flotante': [f"{curva_flotante[0]*100:.4f}%", f"{curva_flotante[1]*100:.4f}%"],
            'Factor Descuento': [f"{swap.factores_descuento[0]:.6f}", f"{swap.factores_descuento[1]:.6f}"]
        })
        pdf.add_table(tabla_swap, "Tabla 6.1: Parametros del swap")

        # An√°lisis num√©rico
        analisis_swap = f"""
        Interpretacion:
        * El valor del swap es ${valor_swap_result:,.2f}, indicando que {'el pagador fijo esta en desventaja' if valor_swap_result > 0 else 'el pagador flotante esta en desventaja'}.
        * El DV01 estimado es de ${(ValoradorSwap(notional, curva_fija, [r+0.0001 for r in curva_flotante], tiempos).valor_swap() - valor_swap_result):,.2f} por cada punto basico de cambio en la curva.

        Limitaciones:
        * El modelo asume que no hay riesgo de credito o de contraparte.
        * Los factores de descuento se calculan utilizando una curva plana.
        """
        pdf.add_markdown(analisis_swap)

        # Gr√°ficos
        fig, axs = plt.subplots(1, 2, figsize=(14, 7))

        # Sensibilidad
        shifts = np.linspace(-0.02, 0.02, 50)
        valores_swap = []
        for shift in shifts:
            curva_shifted = [r + shift for r in curva_flotante]
            swap_shifted = ValoradorSwap(notional, curva_fija, curva_shifted, tiempos)
            valores_swap.append(swap_shifted.valor_swap())
        axs[0].plot(shifts*10000, valores_swap, linewidth=2.5, color=usach_colors[0])
        axs[0].axvline(x=0, color='r', linestyle='--', alpha=0.7)
        axs[0].set_title("Sensibilidad del Swap a Movimientos de la Curva", fontweight='bold')
        axs[0].set_xlabel("Cambio en la Curva (bps)")
        axs[0].set_ylabel("Valor del Swap ($)")
        axs[0].grid(True, alpha=0.3)

        # Curva cero
        axs[1].plot(tiempos, [r*100 for r in curva_flotante], 'o-', linewidth=2.5, color=usach_colors[1])
        axs[1].set_title("Curva Cero Bootstrapped", fontweight='bold')
        axs[1].set_xlabel("Tiempo (a√±os)")
        axs[1].set_ylabel("Tasa (%)")
        axs[1].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.savefig("swap_analysis.png")
        plt.close()

        pdf.add_figure("swap_analysis.png", "Figura 6.1: Sensibilidad del swap y curva de tasas")

    except Exception as e:
        print(f"Error en seccion de swaps: {str(e)}")
        traceback.print_exc()
        pdf.add_markdown("Error en la seccion de swaps. Ver logs para detalles.")

    # =========================================================================
    # 7. VALORACI√ìN DE FRAs
    # =========================================================================
    try:
        print_header("7. valoracion de fras", level=1)
        pdf.add_section("Valoracion de FRAs (6x12) Mejorada", level=1)

        # Markdown explicativo
        fra_insight = """
        Que mirar
        * Valor vs bps - relacion perfectamente lineal (convexidad ‚âà 0).
        * DV01 -$48,9 -> por cada bp de subida pierdes $48,9 por millon de nocional.

        Idea clave
        Util para cubrir un tramo concreto de la curva sin compromisos a mas largo plazo.
        """
        pdf.add_key_insight(fra_insight)

        # Valoraci√≥n de FRA
        fra_params = {
            'notional': 1_000_000,
            'tasa_fra': 0.05,
            'tasa_mercado': 0.048,
            'tiempo': 0.5
        }
        fra = ValoradorFRA(**fra_params)
        valor_fra = fra.valor()

        # Tabla de resultados
        tabla_fra = pd.DataFrame({
            'Parametro': ['Principal Nocional', 'Tasa FRA', 'Tasa Mercado', 'Periodo (a√±os)', 'Valor FRA'],
            'Valor': [f"${fra_params['notional']:,.0f}",
                      f"{fra_params['tasa_fra']*100:.4f}%",
                      f"{fra_params['tasa_mercado']*100:.4f}%",
                      f"{fra_params['tiempo']:.4f}",
                      f"${valor_fra:,.2f}"]
        })
        pdf.add_table(tabla_fra, "Tabla 7.1: Resultados de valoracion de FRA")

        # An√°lisis num√©rico
        analisis_fra = f"""
        Interpretacion:
        * El valor del FRA es ${valor_fra:,.2f}, {'favorable para el comprador' if valor_fra > 0 else 'favorable para el vendedor'}.
        * El DV01 es de ${ValoradorFRA(fra_params['notional'], fra_params['tasa_fra'], fra_params['tasa_mercado']+0.0001, fra_params['tiempo']).valor() - valor_fra:,.2f} por cada punto basico de cambio en la tasa de mercado.

        Limitaciones:
        * El modelo no considera el riesgo de credito o liquidez.
        * Asume que la tasa de mercado es constante hasta el vencimiento.
        """
        pdf.add_markdown(analisis_fra)

        # Gr√°fico de sensibilidad
        shifts_fra = np.linspace(-0.02, 0.02, 50)
        valores_fra = [ValoradorFRA(fra_params['notional'], fra_params['tasa_fra'],
                       fra_params['tasa_mercado'] + s, fra_params['tiempo']).valor() for s in shifts_fra]
        plt.figure(figsize=(10, 6))
        plt.plot(shifts_fra*10000, valores_fra, linewidth=2.5, color=usach_colors[1])
        plt.axvline(x=0, color='r', linestyle='--', alpha=0.7)
        plt.title("Sensibilidad del FRA a Movimientos de Tasas", fontweight='bold')
        plt.xlabel("Cambio en Tasa de Mercado (bps)")
        plt.ylabel("Valor del FRA ($)")
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig("fra_analysis.png")
        plt.close()

        pdf.add_figure("fra_analysis.png", "Figura 7.1: Sensibilidad del FRA a cambios en las tasas")

    except Exception as e:
        print(f"Error en seccion de FRAs: {str(e)}")
        traceback.print_exc()
        pdf.add_markdown("Error en la seccion de FRAs. Ver logs para detalles.")

    # =========================================================================
    # 8. CONCLUSIONES Y RESUMEN
    # =========================================================================
    try:
        print_header("8. conclusiones y resumen", level=1)
        pdf.add_section("Conclusiones y Recomendaciones", level=1)

        # Funci√≥n para formatear valores o devolver "N/A" si no existen
        def format_valor(valor, formato='${:,.2f}'):
            if valor is None:
                return "N/A"
            try:
                return formato.format(valor)
            except:
                return "N/A"

        # Resumen comparativo
        resumen_general = pd.DataFrame({
            'Activo': ['Bono', 'Accion (MSFT)', 'Futuro Petroleo', 'Forward Oro', 'Opcion Call', 'Swap', 'FRA'],
            'Valor': [format_valor(precio_bono),
                     format_valor(valor_ddm),
                     format_valor(futuro_teorico),
                     format_valor(valor_contrato),
                     format_valor(precio_call),
                     format_valor(valor_swap_result, '${:,.2f}'),
                     format_valor(valor_fra, '${:,.2f}')],
            'Rentabilidad': [f"{bono_params['tasa_cupon']*100:.1f}% (cupon)" if 'bono_params' in locals() else "N/A",
                             f"{retorno_anual*100:.2f}%" if retorno_anual is not None else "N/A",
                             'N/A',
                             'N/A',
                             'Variable (apalanc.)',
                             'Spread 0.4%',
                             'Diferencial 0.2%'],
            'Riesgo': [f'Duracion {duracion:.2f} a√±os' if duracion is not None else "N/A",
                       f'Volat. {volatilidad_anual*100:.2f}%' if volatilidad_anual is not None else "N/A",
                       'Precio spot',
                       'Contraparte',
                       'Alto (apalanc.)',
                       'Tasa interes',
                       'Tasa mercado']
        })
        pdf.add_table(resumen_general, "Tabla 8.1: Resumen comparativo de activos")

        # Conclusiones
        conclusiones = """
        Conclusiones y Recomendaciones

        ‚Ä¢ Diversificacion esencial: cada activo ofrece ventajas y sus riesgos son complementarios.
        ‚Ä¢ Bonos: adecuados para perfiles conservadores ‚Üí protegen contra caidas de tasas (alta convexidad).
        ‚Ä¢ Acciones (MSFT): mayor retorno con volatilidad elevada ‚Üí inversor tolerante al riesgo.
        ‚Ä¢ Futuros & Forwards: cobertura de commodities ‚Üí ideales para gestion de inventarios.
        ‚Ä¢ Opciones: apalancamiento puro ‚Üí traders de volatilidad.
        ‚Ä¢ Swaps & FRAs: ajuste fino de brechas de tasas (DV01) ‚Üí tesorerias y ALM.

        Recomendaciones por perfil:
        - Conservador: Bonos mediana duracion + Swap corto DV01 bajo.
        - Moderado: Acciones blue-chip + futuros de materias primas.
        - Agresivo: Opciones y estrategias de volatilidad.
        """
        pdf.add_markdown(conclusiones)

        # Gr√°fico de retorno/riesgo
        try:
            fig, ax = plt.subplots(figsize=(8, 4))
            activos = ['Bonos', 'Acciones', 'Futuros', 'Opciones', 'Swap', 'FRA']
            retorno = [0.05, 0.12, 0.08, 0.25, 0.04, 0.03]
            riesgo = [0.03, 0.18, 0.12, 0.40, 0.05, 0.04]

            colors = usach_colors * 2  # Extender la paleta si es necesario
            for i in range(len(activos)):
                ax.scatter(riesgo[i], retorno[i], s=100, c=colors[i])
                ax.annotate(activos[i], (riesgo[i], retorno[i]), xytext=(10, 5),
                             textcoords='offset points', fontsize=9)

            ax.set_title("Retorno vs Riesgo por Clase de Activo", fontweight='bold')
            ax.set_xlabel("Riesgo (Volatilidad)")
            ax.set_ylabel("Retorno Esperado")
            ax.grid(True, alpha=0.3)
            plt.tight_layout()
            plt.savefig("retorno_riesgo.png")
            pdf.add_figure("retorno_riesgo.png", "Figura 8.1: Relacion retorno-riesgo por activo")
        except:
            pdf.add_markdown("‚ö† Grafico de retorno/riesgo no disponible")

    except Exception as e:
        print(f"Error en seccion de conclusiones: {str(e)}")
        traceback.print_exc()
        pdf.add_markdown("Error en la seccion de conclusiones. Ver logs para detalles.")

    # =========================================================================
    # 9. AP√âNDICES Y BIBLIOGRAF√çA
    # =========================================================================
    pdf.add_appendix()
    pdf.add_bibliography()

    # Guardar PDF
    pdf.output('Informe_Valoracion_USACH.pdf')
    print("Informe generado exitosamente: Informe_Valoracion_USACH.pdf")

if __name__ == "__main__":
    main()
```
